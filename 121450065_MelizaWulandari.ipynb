```python
import pandas as pd
from sklearn.cluster import KMeans

# Data mahasiswa
data = {
    'Nama Mahasiswa': ['Roy', 'Sintia', 'Iqbal', 'Dilan', 'Ratna',
                       'Merry', 'Rudi', 'Hafiz', 'Gede', 'Christian',
                       'Justin', 'Jesika', 'Ayu', 'Siska', 'Reitama'],
    'Nilai UTS': [89, 90, 70, 45, 65, 80, 90, 70, 96, 60, 45, 60, 85, 52, 40],
    'Nilai Tugas': [90, 71, 75, 65, 75, 70, 85, 70, 93, 55, 60, 70, 90, 68, 60],
    'Nilai UAS': [75, 95, 80, 59, 53, 75, 81, 73, 85, 48, 58, 72, 88, 55, 70]
}

df = pd.DataFrame(data)

# Menggabungkan nilai-nilai
X = df[['Nilai UTS', 'Nilai Tugas', 'Nilai UAS']]

# Menggunakan algoritma k-means untuk klasterisasi menjadi 3 kelompok
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

# Menambahkan kolom 'Klaster' ke DataFrame
df['Klaster'] = kmeans.labels_

df['Klaster'] = df['Klaster'] + 1

# Output DataFrame
print(df)
```

Analisis kode

kode diatas menggunakan K-Means untuk mengelompokkan data kedalam 3 klaster (n_clusters=3) yaitu dengan mengelompokkan mahasiswa ke dalam tiga kelompok berdasarkan nilai UTS, Tugas, dan UAS.



Analisis hasil

mahasiswa dengan nilai UTS, Tugas dan UAS yang tinggi ditempatkan dalam klaster 3 yang termasuk kedalam kategori mahasiswa yang berprestasi .
mahasiswa dengan nilai UTS, Tugas dan UAS yang berada di tengah-tengah ditempatkan dalam klaster 2 yang termasuk kedalam kategori mahasiswa yang sedang.
mahasiswa dengan nilai UTS, Tugas dan UAS yang relatif kecil ditempatkan dalam klaster 1 yang termasuk kedalam kategori mahasiswa yang rendah.
