```python
import numpy as np

# menginputkan data input(x) dan target(t)
x = np.array([1, 0, -1])
t = np.array([1, 0, 2])

#fungsi basis gausian p0(x)=1 dan p1(x)=exp(-x^2/2)
phi_0 = np.ones(x.shape)
phi_1 = np.exp(-x**2/2)

# matriks desain
X = np.column_stack((phi_0, phi_1))

# menghitung vektor parameter (koefisien)
w = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(t)

# solusi regresi linier
print("koefisien w:", w)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Menginputkan Data
data = np.array([[1, 1], [0, 0], [-1, 2]])

# Fungsi basis Gaussian p0(x)=1 dan p1(x)=exp(-x^2/2)
def basis_functions(x):
    return np.array([1, np.exp(-x**2/2)])

# Membuat matriks desain
Phi = np.array([basis_functions(x) for x in data[:, 0]])

# Vektor target
t = data[:, 1]

# Menghitung vektor target (w)
w = np.linalg.inv(Phi.T @ Phi) @ Phi.T @ t

# Menampilkan vektor bobot
print("Vektor targetnya (w):", w)
```

```python
# Membuat prediksi pada sejumlah titik data
x_values = np.linspace(-2, 2, 400)  # Nilai x dari -2 hingga 2
predicted_values = np.array([basis_functions(x) @ w for x in x_values])

# Plot data asli
plt.scatter(data[:, 0], data[:, 1], color='red', label='Data Asli')

# Plot hasil prediksi
plt.plot(x_values, predicted_values, color='blue', label='Regresi Bayesian')

# Menambahkan label dan legenda
plt.xlabel('x')
plt.ylabel('t')
plt.legend()

# Menampilkan plot
plt.show()
```
