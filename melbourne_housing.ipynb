```python
import matplotlib.pyplot as plt
import seaborn as sns

# Menghitung rata-rata harga properti untuk setiap agen real estate
avg_price_by_seller = df.groupby('SellerG')['Price'].mean().sort_values(ascending=False).head(10)  # Top 10 sellers by price

# Membuat bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_price_by_seller.index, y=avg_price_by_seller.values, palette='viridis')

# Menambahkan judul dan label
plt.title('Rata-rata Harga Properti Berdasarkan Agen Real Estate (Top 10)', fontsize=16)
plt.xlabel('Agen Real Estate', fontsize=12)
plt.ylabel('Rata-rata Harga Properti (AUD)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Menampilkan chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Menghitung distribusi SellerG
seller_counts = df['SellerG'].value_counts()

# Membuat pie chart
plt.figure(figsize=(10, 7))
plt.pie(seller_counts, labels=seller_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Distribusi SellerG (Agen Real Estate)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Menampilkan chart
plt.show()
```
