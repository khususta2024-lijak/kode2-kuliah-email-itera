```python
# Pemrograman Prosedural dengan 3 Fungsi
def f(x):
  return 4 ** x

def g(x):
  return x ** x

def h(x):
  return 3 + x

def komposisi_prosedural(x):
  return f(g(h(x)))
```

```python
print(komposisi_prosedural(2))
```

```python
# Pemrograman Fungsional dengan Lambda
f = lambda x: 4 ** x
g = lambda x: x ** x
h = lambda x: 3 + x

komposisi_fungsional = lambda x: f(g(h(x)))
```

```python
print(komposisi_fungsional(2))
```

# Nomor 2
### Buatlah pemrograman berbasis objek pada soal no. 1

```python
class Fungsi:
  def __init__(self, rumus):
    self.rumus = rumus

  def hitung(self, x):
    return self.rumus(x)

class Komposisi:
  def __init__(self, f, g, h):
    self.f = f
    self.g = g
    self.h = h

  def hitung(self, x):
    return self.f.hitung(self.g.hitung(self.h.hitung(x)))


# Membuat objek dari kelas fungsi untuk masing-masing fungsi f(x), g(x), dan h(x)
f = Fungsi(lambda x: 4 ** x)
g = Fungsi(lambda x: x ** x)
h = Fungsi(lambda x: 3 + x)

# Membuat objek dari kelas komposisi dengan fungsi-fungsi yang sudah dibuat
komposisi = Komposisi(f, g, h)

# Menggunakan objek komposisi untuk menghitung hasil fungsi komposisi dari x = 2
print(komposisi.hitung(2))
```

### Berikan Analisis Komparasi soal 1 dan 2

Soal 1 dan 2 sama-sama membahas tentang penghitungan fungsi komposisi dari tiga fungsi matematika yang diberikan, yaitu f(x) = 4^x,
g(x) = x^x, dan h(x) = 3 + x.

Perbedaan utama antara kedua soal tersebut terletak pada paradigma pemrograman yang digunakan dalam implementasinya. Soal nomor 1 meminta implementasi program dengan paradigma pemrograman prosedural dan fungsional dengan lambda function, sedangkan Soal nomor 2 meminta implementasi program dengan paradigma pemrograman berbasis objek.

Implementasi program dengan paradigma pemrograman prosedural dan fungsional pada Soal 1 lebih cocok digunakan dalam
implementasi program dengan fitur-fitur yang relatif sederhana, sedangkan implementasi program dengan paradigma pemrograman
berbasis objek pada Soal 2 lebih cocok digunakan dalam implementasi program yang lebih kompleks dan memiliki struktur yang lebih
kompleks pula. Dalam paradigma pemrograman berbasis objek, kode program dibagi ke dalam beberapa kelas dan objek, yang
membantu untuk mengorganisir dan memahami struktur program secara keseluruhan.
Pada akhirnya, pemilihan antara paradigma pemrograman prosedural dan fungsional atau paradigma pemrograman berbasis objek akan
tergantung pada konteks program yang akan diimplementasikan. Jika program yang dibuat memiliki fitur-fitur yang relatif sederhana,
paradigma pemrograman prosedural dan fungsional dapat menjadi pilihan yang tepat. Namun, jika program yang dibuat lebih kompleks
dan memiliki struktur yang lebih kompleks pula, paradigma pemrograman berbasis objek dapat menjadi pilihan yang lebih baik.
