```python
# Membership functions untuk kinerja
kinerja['rendah'] = fuzz.trapmf(kinerja.universe, [0, 0, 50, 50])
kinerja['sedang'] = fuzz.trapmf(kinerja.universe, [50, 50, 80, 80])
kinerja['tinggi'] = fuzz.trapmf(kinerja.universe, [80, 80, 100, 100])
```

```python
# Membership functions untuk pegawai terbaik
pegawai_terbaik['bukan terbaik'] = fuzz.trapmf(pegawai_terbaik.universe, [0, 0, 85, 85])
pegawai_terbaik['terbaik'] = fuzz.trapmf(pegawai_terbaik.universe, [85, 85, 100, 100])
```

```python
# Aturan-aturan fuzzy
rule1 = ctrl.Rule(kedisiplinan['cukup'] & kinerja['rendah'], pegawai_terbaik['bukan terbaik'])
rule2 = ctrl.Rule(kedisiplinan['cukup'] & kinerja['sedang'], pegawai_terbaik['bukan terbaik'])
rule3 = ctrl.Rule(kedisiplinan['cukup'] & kinerja['tinggi'], pegawai_terbaik['bukan terbaik'])
rule4 = ctrl.Rule(kedisiplinan['sangat baik'] & kinerja['rendah'], pegawai_terbaik['bukan terbaik'])
rule5 = ctrl.Rule(kedisiplinan['sangat baik'] & kinerja['sedang'], pegawai_terbaik['terbaik'])
rule6 = ctrl.Rule(kedisiplinan['sangat baik'] & kinerja['tinggi'], pegawai_terbaik['terbaik'])
```

```python
# Sistem kontrol fuzzy
pegawai_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6])
pegawai_evaluation = ctrl.ControlSystemSimulation(pegawai_ctrl)
```

```python
# Menambahkan kolom hasil fuzzy
df['Skor Pegawai Terbaik'] = 0
```

```python
# Evaluasi setiap pegawai
for index, row in df.iterrows():
    pegawai_evaluation.input['kedisiplinan'] = row['kedisiplinan']
    pegawai_evaluation.input['kinerja'] = row['rata-rata kinerja']
    pegawai_evaluation.compute()
    df.at[index, 'Skor Pegawai Terbaik'] = pegawai_evaluation.output['pegawai_terbaik']
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# Definisikan variabel fuzzy untuk kedisiplinan, kinerja, dan kerjasama
kedisiplinan = ctrl.Antecedent(np.arange(0, 101, 1), 'kedisiplinan')
kinerja = ctrl.Antecedent(np.arange(0, 101, 1), 'kinerja')
kerjasama = ctrl.Antecedent(np.arange(0, 101, 1), 'kerjasama')
pegawai_terbaik = ctrl.Consequent(np.arange(0, 101, 1), 'pegawai_terbaik')

# Membership functions untuk kedisiplinan
kedisiplinan['cukup'] = fuzz.trapmf(kedisiplinan.universe, [0, 0, 50, 75])
kedisiplinan['sangat baik'] = fuzz.trapmf(kedisiplinan.universe, [75, 100, 100, 100])

# Membership functions untuk kinerja
kinerja['rendah'] = fuzz.trapmf(kinerja.universe, [0, 0, 50, 60])
kinerja['sedang'] = fuzz.trimf(kinerja.universe, [50, 70, 80])
kinerja['tinggi'] = fuzz.trapmf(kinerja.universe, [70, 80, 100, 100])

# Membership functions untuk kerjasama
kerjasama['rendah'] = fuzz.trapmf(kerjasama.universe, [0, 0, 50, 60])
kerjasama['sedang'] = fuzz.trimf(kerjasama.universe, [50, 70, 80])
kerjasama['tinggi'] = fuzz.trapmf(kerjasama.universe, [70, 80, 100, 100])

# Membership functions untuk pegawai terbaik
pegawai_terbaik['tidak'] = fuzz.trimf(pegawai_terbaik.universe, [0, 0, 50])
pegawai_terbaik['ya'] = fuzz.trimf(pegawai_terbaik.universe, [50, 100, 100])

# Aturan fuzzy
rule1 = ctrl.Rule(kedisiplinan['cukup'] & kinerja['rendah'] & kerjasama['rendah'], pegawai_terbaik['tidak'])
rule2 = ctrl.Rule(kedisiplinan['cukup'] & kinerja['sedang'] & kerjasama['rendah'], pegawai_terbaik['tidak'])
rule3 = ctrl.Rule(kedisiplinan['cukup'] & kinerja['tinggi'] & kerjasama['rendah'], pegawai_terbaik['tidak'])
rule4 = ctrl.Rule(kedisiplinan['cukup'] & kinerja['rendah'] & kerjasama['sedang'], pegawai_terbaik['tidak'])
rule5 = ctrl.Rule(kedisiplinan['cukup'] & kinerja['sedang'] & kerjasama['sedang'], pegawai_terbaik['tidak'])
rule6 = ctrl.Rule(kedisiplinan['cukup'] & kinerja['tinggi'] & kerjasama['sedang'], pegawai_terbaik['tidak'])
rule7 = ctrl.Rule(kedisiplinan['cukup'] & kinerja['rendah'] & kerjasama['tinggi'], pegawai_terbaik['tidak'])
rule8 = ctrl.Rule(kedisiplinan['cukup'] & kinerja['sedang'] & kerjasama['tinggi'], pegawai_terbaik['tidak'])
rule9 = ctrl.Rule(kedisiplinan['cukup'] & kinerja['tinggi'] & kerjasama['tinggi'], pegawai_terbaik['tidak'])

rule10 = ctrl.Rule(kedisiplinan['sangat baik'] & kinerja['rendah'] & kerjasama['rendah'], pegawai_terbaik['tidak'])
rule11 = ctrl.Rule(kedisiplinan['sangat baik'] & kinerja['sedang'] & kerjasama['rendah'], pegawai_terbaik['tidak'])
rule12 = ctrl.Rule(kedisiplinan['sangat baik'] & kinerja['tinggi'] & kerjasama['rendah'], pegawai_terbaik['tidak'])
rule13 = ctrl.Rule(kedisiplinan['sangat baik'] & kinerja['rendah'] & kerjasama['sedang'], pegawai_terbaik['tidak'])
rule14 = ctrl.Rule(kedisiplinan['sangat baik'] & kinerja['sedang'] & kerjasama['sedang'], pegawai_terbaik['ya'])
rule15 = ctrl.Rule(kedisiplinan['sangat baik'] & kinerja['tinggi'] & kerjasama['sedang'], pegawai_terbaik['ya'])
rule16 = ctrl.Rule(kedisiplinan['sangat baik'] & kinerja['rendah'] & kerjasama['tinggi'], pegawai_terbaik['tidak'])
rule17 = ctrl.Rule(kedisiplinan['sangat baik'] & kinerja['sedang'] & kerjasama['tinggi'], pegawai_terbaik['ya'])
rule18 = ctrl.Rule(kedisiplinan['sangat baik'] & kinerja['tinggi'] & kerjasama['tinggi'], pegawai_terbaik['ya'])

# Sistem kontrol
pegawai_terbaik_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9,
                                           rule10, rule11, rule12, rule13, rule14, rule15, rule16, rule17, rule18])
pegawai_terbaik_simulasi = ctrl.ControlSystemSimulation(pegawai_terbaik_ctrl)

# Data pegawai
data_pegawai = [
    {"nama": "Ari Rusmasari SST, M.Si.", "kedisiplinan": 100, "kinerja": 92.743333, "kerjasama": 220.142857},
    {"nama": "Niken Hariyanti S.Si.", "kedisiplinan": 100, "kinerja": 92.226667, "kerjasama": 211.571429},
    {"nama": "Erika Haryulistiani Saksono SE.", "kedisiplinan": 100, "kinerja": 92.323333, "kerjasama": 221.000000}
]

# Simulasi dan evaluasi pegawai terbaik
for pegawai in data_pegawai:
    pegawai_terbaik_simulasi.input['kedisiplinan'] = pegawai['kedisiplinan']
    pegawai_terbaik_simulasi.input['kinerja'] = pegawai['kinerja']
    pegawai_terbaik_simulasi.input['kerjasama'] = pegawai['kerjasama']

    pegawai_terbaik_simulasi.compute()

    pegawai['nilai_pegawai_terbaik'] = pegawai_terbaik_simulasi.output['pegawai_terbaik']
    print(f"Nama Pegawai: {pegawai['nama']}, Nilai Pegawai Terbaik: {pegawai['nilai_pegawai_terbaik']}")

```
