```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (16, 12)
plt.rcParams.update({'font.size':18})
```

```python
# membuat sinyal sederhana dengan 2 komponen frekuensi
dt = 0.001
t = np.arange(0, 1, dt) # mulai dari 0 sampai 1 dengan spasi 0.001
f = np.sin(2*np.pi*50*t) # frekuensi nya 50hz
f_clean = f # frekuensi bersihnya
f = f + 2.5*np.random.randn(len(t)) # menambahkan noise. noise lebih besar daripada sinyal asli
```

```python
# membuat visualisasi sinyal tanpa noise
fig,axs = plt.subplots(2,1)

plt.sca(axs[0])
plt.plot(t, f_clean, color = 'k', linewidth = 2.5, label = 'clean') # Changed LineWidth to linewidth
plt.xlim(t[0], t[-1])
plt.ylim(-7.5, 7.5)
plt.xlabel('time (s)')
plt.ylabel('amplitude')
plt.legend()

# membuat visualisasi sinyal mempunyai noise
plt.sca(axs[1])
plt.plot(t, f, color = 'k', linewidth = 1.5, label = 'noisy')
plt.xlim(t[0], t[-1])
plt.ylim(-7.5, 7.5)
plt.xlabel('time (s)')
plt.ylabel('amplitude')
plt.legend()

plt.show()
```

```python
# menghitung FFT
n = len(t) # jumlah sampel sama dengan panjang t nya
fhat = np.fft.fft(f, n) # menghitung FFT
PSD = fhat * np.conj(fhat) / n # power spectrum density (representasi lain dari hasil FFT)
freq = (1 / (dt * n)) * np.arange(n) # membuat sumbu x frekuensi
L = np.arange(1, np.floor(n / 2), dtype = 'int') # hanya mengambil setengah data
```

```python
# menyatukan sinyal yang tidak ada noise dan ada noise
fig, axs = plt.subplots(2, 1)

plt.sca(axs[0])
plt.plot(t, f, color = 'c', linewidth = 1.5, label = 'noisy')
plt.plot(t, f_clean, color = 'k', linewidth = 2.5, label = 'clean')
plt.xlim(t[0], t[-1])
plt.ylim(-7.5, 7.5)
plt.xlabel('time (s)')
plt.ylabel('amplitude')
plt.legend()
```

```python
plt.sca(axs[1])
plt.plot(freq[L], PSD[L], color = 'c', linewidth = 2, label = 'PSD')
plt.xlim(freq[L[0]], freq[L[-1]])
plt.xlabel('frekuensi')
plt.ylabel('power')
plt.legend()
```

```python
plt.show()
```

frekuensi paling menonjol adalah frekuensi 50. karena saat membuat sinyal frekuensinya kita gunakan 50. f = np.sin(2*np.pi*50*t) # frekuensi nya 50hz

```python
#menggunakan hasil FFT untuk menghilangkan noise
indices = PSD > 100 # menemukan semua frekuensi dengan power > 100
PSDclean = PSD * indices # mengenolkan frekuensi yang lain
fhat = indices * fhat # fhat hanya berisi frekuensi yang besar
ffilt = np.fft.ifft(fhat) # invers FFT untuk mengembalikan ke domain waktu
```

```python
# plot sinyal noise dan clean
fig, axs = plt.subplots(3,1)

plt.sca(axs[0])
plt.plot(t, f, color = 'c', linewidth = 1.5, label = 'noisy')
plt.plot(t, f_clean, color = 'k', linewidth = 2.5, label = 'clean')
plt.xlim(t[0], t[-1])
plt.xlabel('time (s)')
plt.ylabel('amplitude')
plt.legend()
```

```python
# hasil fft
plt.sca(axs[1])
plt.plot(freq[L], PSD[L], color = 'c', linewidth = 2, label = 'PSD')
plt.xlim(freq[L[0]], freq[L[-1]])
plt.xlabel('frekuensi')
plt.ylabel('power')
plt.legend()
```

```python
# hasil invers fft
plt.sca(axs[2])
plt.plot(t, ffilt, color = 'k', linewidth = 2, label = 'filtered')
plt.xlim(t[0], t[-1])
plt.xlabel('time (s)')
plt.ylabel('amplitude')
plt.legend()

plt.show()
```
