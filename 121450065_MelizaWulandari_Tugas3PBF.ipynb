##1. Ubah fungsiku menjadi pure function

````python
def fungsiku(L, n):
    def check_genap(l):
        return l % 2 == 0
    
    result = []
    for i in range(len(L)):
        if check_genap(L[i]):
            result.append(L[i] / 2)
        else:
            result.append(L[i] * n + 1)
    
    return result

n = 3
L = [5, 6, 7, 8]

L_modified = fungsiku(L, n)
print(L_modified)

print(L)
````

##2. Ubah fungsiku2 menjadi pure function

````python
def fungsiku2(L, n):
    def check_faktor(l):
        return l % n == 0
    
    result = []
    for i in range(len(L)):
        if check_faktor(L[i]):
            result.append(L[i] / 2)
        else:
            result.append(L[i] * n + 1)
    
    return result

n = 3
L = [5, 6, 7, 8]

L_modified = fungsiku2(L, n)
print(L_modified)

print(L)
````

##5. Penentuan UKT Mahasiswa

Universitas di Lampung ITERA, ingin memiliki sistem penentuan golongan UKT dan jumlah biaya UKT yang dibayarkan oleh Mahasiswa berdasarkan Kriteria berikut:
1. Jumlah tanggungan
2. Jumlah token listrik selama 3 bulan terakhir
3. Gaji Orang tua / Penanggung jawab
4. Penerima program KIP-K atau bukan



1. Ketentuan Jumlah Tanggungan :\
   Jika lebih >= 5, maka skor = 1\
   Jika < 5 , maka skor = 5 - jumlah tanggungan



2. Ketentuan token listrik:\
   Jika rata-rata lebih dari 100 ribu per bulan , maka skor = 3\
   Jika diantara 50 ribu - 100 ribu per bulan , maka skor = 2\
   Jika dibawah 50 ribu , maka skor = 1



3. Ketentuan Gaji :\
   Jika gaji penanggung jawab > 10 juta maka skor = 7\
   Jika 8 < gaji <= 10 juta, maka skor = 6\
   Jika 6 < gaji <= 8 juta , maka skor = 5\
   Jika 4 < gaji <= 6 juta , maka skor = 4\
   Jika 3< gaji <= 4 juta , maka skor = 3\

4. Jika mahasiswa memiliki KIP-K , maka skor = 1, jika tidak maka\
skor = 5\
Perhitungan pembayaran UKT adalah sebagai berikut:\
Skor_total = 20 % * skor_1 + 30 % * skor_2 + 20% skor_3 + 30%\
skor_4\
Jumlah bayar UKT = biaya pokok + skor_total * 500 ribu\
Uang Pokok = 750 ribu\
Gunakan fungsi komposisi untuk menyelesaikan masalah tersebut!\
Hitung berapa biaya UKT yang harus dibayarkan dengan input\
sebagai berikut:\
Jumlah tanggungan = 3\
Listrik 3 bulan terakhir = 120 ribu , 75 ribu , 50 ribu\
Gaji Penanggung jawab = 5.5 juta per bulan\
Peserta KIP-K = Tidak


````python
def penentuan_ukt(jumlah_tanggungan, token_listrik, gaji, kipk):
    # Komposisi fungsi untuk menghitung skor berdasarkan kriteria
    skor_1 = lambda tanggungan: 1 if tanggungan >= 5 else 5 - tanggungan
    skor_2 = lambda rata_listrik: 3 if rata_listrik > 100000 else (2 if 50000 <= rata_listrik <= 100000 else 1)
    skor_3 = lambda gaji: (
        7 if gaji > 10000000 else
        6 if 8000000 < gaji <= 10000000 else
        5 if 6000000 < gaji <= 8000000 else
        4 if 4000000 < gaji <= 6000000 else
        3 if 3000000 < gaji <= 4000000 else
        2
    )
    skor_4 = lambda kipk: 1 if kipk == "Ya" else 5
    
    # Menghitung skor berdasarkan kriteria
    skor_total = (
        0.2 * skor_1(jumlah_tanggungan) +
        0.3 * skor_2(sum(token_listrik) / len(token_listrik)) +
        0.2 * skor_3(gaji) +
        0.3 * skor_4(kipk)
    )
    
    # Menghitung jumlah bayar UKT
    biaya_pokok = 750000
    jumlah_bayar_ukt = biaya_pokok + skor_total * 500000
    
    return jumlah_bayar_ukt

jumlah_tanggungan = 3
token_listrik = [120000, 75000, 50000]
gaji = 5500000
kipk = "Tidak"

biaya_ukt = penentuan_ukt(jumlah_tanggungan, token_listrik, gaji, kipk)
print("Biaya UKT yang harus dibayarkan:", biaya_ukt)
````

##6 Turunan Polinom

contoh input:


```dat = '-3x^5 ++ 2x^2 - 4x + 5'```


output:

'-15.0x^4 + 4.0x -4.0'

````python
import re

def turunan_polinom(polinom):
    terms = re.findall(r'(-?\d*)x\^(\d+)', polinom)
    turunan_terms = []
    
    for koefisien, eksponen in terms:
        if koefisien == '':
            koefisien = '1'
        
        if eksponen == '1':
            turunan_terms.append(koefisien)
        else:
            koefisien_baru = str(float(koefisien) * int(eksponen))
            eksponen_baru = str(int(eksponen) - 1)
            turunan_terms.append(f"{koefisien_baru}x^{eksponen_baru}")
    
    turunan = ' + '.join(turunan_terms)
    return turunan

dat = '-3x^5 ++ 2x^2 - 4x + 5'
hasil = turunan_polinom(dat)
print(hasil)
````

##7 Buatlah fungsi untuk menghitung biaya yang harus dibayar customer pada suatu e-commerce menggunakan higher order function. Buatlah decorator untuk mengeluarkan harga sebelum pajak dan sesudah pajak (pajak = 11%) ! Gunakan decorator untuk menambahkan perhitungan waktu eksekusi!

keranjang = [\
  {'Jumlah_Barang': 5 , 'Harga': 10},\
  {'Jumlah_Barang': 7 , 'Harga': 20},\
  {'Jumlah_Barang': 20 , 'Harga': 4.5}\
]

````python
import time

# Higher order function untuk menghitung biaya pembelian
def hitung_biaya(pajak):
    def decorator(func):
        def wrapper(keranjang):
            total_harga = sum(item['Jumlah_Barang'] * item['Harga'] for item in keranjang)
            harga_sebelum_pajak = total_harga
            harga_setelah_pajak = total_harga * (1 + pajak)
            print(f'Harga sebelum pajak: {harga_sebelum_pajak}')
            print(f'Harga setelah pajak: {harga_setelah_pajak}')
            return func(keranjang)
        return wrapper
    return decorator

# Decorator untuk menghitung waktu eksekusi
def hitung_waktu(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        execution_time = end_time - start_time
        print(f'Waktu eksekusi: {execution_time} detik')
        return result
    return wrapper

# Fungsi untuk menghitung biaya pembelian
@hitung_biaya(pajak=0.11)
@hitung_waktu
def hitung_total_biaya(keranjang):
    return sum(item['Jumlah_Barang'] * item['Harga'] for item in keranjang)

# Data keranjang belanja
keranjang = [
    {'Jumlah_Barang': 5, 'Harga': 10},
    {'Jumlah_Barang': 7, 'Harga': 20},
    {'Jumlah_Barang': 20, 'Harga': 4.5}
]

# Menghitung total biaya dengan menggunakan decorator
total_biaya = hitung_total_biaya(keranjang)
print(f'Total biaya: {total_biaya}')
````
