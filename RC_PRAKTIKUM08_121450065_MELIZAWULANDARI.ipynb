## Fibbonacci Problem
## FIbonacci adalah suatu barisan bilangan yang merupakan hasil penjumlahan dua bilangan sebelumnya.
image.png
## Fibonacci menggunakan perulangan untuk menjalankan perhitungannya lengkapi kode fibonacci dibawah ini

```python
def fib(numbers):
  fibo = [0,1] #Base
  if numbers > 2:
    for i in range (2, numbers):
      #elemen berikutnya dalam seri = jumlah dari dua angka sebelumnya
      nextElement = fibo[i-1] + fibo[i-2]
      #menambahkan elemen ke list
      fibo.append(nextElement)

  return fibo

fib(65)
```

## 3. Buatlah fungsi looping diatas menjadi looping rekursif

```python
def fibonacci(num):
    base = [0, 1]
    if num in base:
      return base[num]
    return fibonacci(num-1) + fibonacci(num-2) #Rekursif Case

[fibonacci(num) for num in range(11)]
# maaf kak, sebenarnya ujung nim saya 65, tetapi disini saya masukkan 11 karena jika 65 proses running nya lama sekali
```

```python
def factorial(n):
  return 1 if n==1 else n*factorial(n-1)
```

## 6. Buatlah sebuah fungsi rekursif untuk mencari bilangan kuadrat ke-n, yaitu bilangan bulat yang merupakan hasil pangkat dua dari n!

```python
def kuadrat(n):
  if n == 0:
    return 0
  else:
    return kuadrat(n-1) + (2*n) - 1
```

```python
kuadrat(5)
```

## 7. Lengkapilah code pada fungsi rekursiff untuk menghitung jumlah elemen dalah sebuah list!

```python
def jumlah_elemen(list):
  if len(list) == 0:
    return 0
  else:
    return 1 + jumlah_elemen(list[1:])

data = [1, 2, 3, 4, 5]
print(jumlah_elemen(data))
```
