4. (10 poin) Lakukan transformasi data dengan mengonversi data kategorikal menjadi data numerik menggunakan Label Encoding pada dataset diabetes_RS2

```python
# Pilih semua kolom yang bukan numerik dengan menggunakan df.select_dtypes(exclude=[np.number]).columns
import numpy as np
df_category = df3.select_dtypes(exclude=[np.number]).columns
```

```python
# tampilkan lima baris pertama dataframe baru
df3[df_category].head(5)
```

```python
# ubah kolom kategorikal menjadi data numerik, dengan cara import paket sklearn.preprocessing dan gunakan LabelEncoder()
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for i in df_category:
  df3[i] = label_encoder.fit_transform(df3[i])
print("hasil transformation: ")
df3.head()
```

```python
# simpan data .csv untuk dataset diabetes_RS1 dan diabetes_RS2 yang baru.
df1.to_csv("df12")
df3.to_csv("data12")
```
