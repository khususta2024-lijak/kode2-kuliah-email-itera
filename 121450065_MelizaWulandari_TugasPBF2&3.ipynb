##1. Buatlah fungsi mencari bilangan genap dari list L!
## L = [2,1,9,10,3,90,15]

```python
from functools import reduce

def jumlah_genap(L):
  # reduce biasa
  hitung1 = reduce(lambda x, y: x + (y % 2 == 0), L, 0)

  # reduce initializer
  hitung2 = reduce(lambda x, y: x + (y % 2 == 0), L)

  return hitung1, hitung2

L = [2,1,9,10,3,90,15]
hitung1, hitung2 = jumlah_genap(L)

print("Menggunakan reduce biasa = ", hitung1)
print("Menggunakan reduce initializer = ", hitung2)
```

Pada kode diatas terdapat fungsi "jumlah_genap" yang menerima sebuah list sebagai parameter. Fungsi tersebut menghitung jumlah bilangan genap pada list menggunakan fungsi reduce dari modul functools.

Pada fungsi "jumlah_genap", terdapat dua penggunaan fungsi reduce. Fungsi reduce pertama menghitung jumlah bilangan genap pada list dengan cara menjumlahkan setiap elemen y pada list L yang bernilai True ketika y modulo 2 sama dengan 0. Sedangkan pada fungsi reduce kedua, parameter ketiga (initializer) dihilangkan dan nilai initial diambil dari nilai pertama pada list L. Sehingga pada fungsi reduce kedua, setiap elemen y pada list L yang bernilai True ketika y modulo 2 sama dengan 0, akan dijumlahkan dengan nilai pertama pada list.

```python
from functools import reduce

def jumlah_ganjil(L):
  # reduce biasa
  hitung1 = reduce(lambda x, y: x + (y % 2 != 0), L, 0)

  # reduce initializer
  hitung2 = reduce(lambda x, y: x + (y % 2 != 0), L)

  return hitung1, hitung2

L = [2,1,9,10,3,90,15]
hitung1, hitung2 = jumlah_ganjil(L)

print("Menggunakan reduce biasa = ", hitung1)
print("Menggunakan reduce initializer = ", hitung2)
```

Pada reduce biasa, menggunakan lambda function untuk menambahkan 1 ke nilai accumulator x jika elemen y dari list l adalah bilangan ganjil (tidak habis dibagi 2), dan mengembalikan nilai accumulator x sebagai jumlah bilangan ganjil pada list l.

Pada reduce initializer, memiliki cara kerja yang sama dengan count_odd_numbers_reduce(), tetapi nilai accumulator x diinisialisasi dengan nilai 0 secara otomatis oleh reduce() sehingga tidak perlu diberikan sebagai argumen terpisah.

## 2. Buatlah fungsi untuk menghitung N menggunakan reduce

```python
from functools import reduce

def menghitung_n(n):
  return reduce(lambda x,y: x+y, range(n+1))
```

```python
print(menghitung_n(4))
```

Kode di atas akan menghasilkan output berupa hasil penjumlahan bilangan bulat dari 0 hingga 4 (inklusif), yaitu 10. Pada saat kode dijalankan, fungsi menghitung_n dipanggil dengan argumen 4. Fungsi ini menggunakan fungsi reduce untuk menjumlahkan semua bilangan bulat dari 0 hingga 4. 
Fungsi reduce menggunakan lambda function untuk menjumlahkan dua bilangan pada setiap iterasi hingga seluruh bilangan bulat dari 0 hingga 4 dijumlahkan. Kemudian hasil penjumlahan, yaitu 10, dicetak menggunakan perintah print.

##3. Hitung euclidian distance dari dua vektor beerikut menggunakan higher order function
## X= [2,5,6,7,10]
## Y= [-2,9,2,-1,10]

```python
from functools import reduce

x = [2, 5, 6, 7, 10]
y = [-2, 9, 2, -1, 10]

def euclidian(x,y):
  return (x-y) **2
euclidian = map(euclidian, x, y)

sum_euclidian = reduce(lambda x,y: x + y, euclidian)

euclidian_distance = sum_euclidian ** 0.5

print("Euclidian distance dari x dan y= ", euclidian_distance)
```

Kode di atas akan menghasilkan output berupa nilai Euclidean distance dari dua list, x dan y. Pertama-tama, kita mendefinisikan fungsi euclidian yang akan digunakan untuk menghitung kuadrat dari selisih antara dua angka. Kemudian, kita menggunakan fungsi map untuk menerapkan fungsi euclidian pada setiap pasangan elemen dari dua list x dan y.

Selanjutnya, kita menggunakan fungsi reduce untuk menjumlahkan hasil kuadrat selisih yang diperoleh dari setiap pasangan elemen. Hasil penjumlahan ini kemudian diambil akar kuadratnya dengan memangkatkan hasil penjumlahan tersebut dengan pangkat 0.5. Akhirnya, hasil Euclidean distance dicetak menggunakan perintah print.

##5. Buatlah deret fibonacci menggunakan higher order function

```python
from functools import reduce

def fibonacci(n):
    # Menggunakan reduce biasa
    fibonacci_biasa = reduce(lambda x, y: x + [x[-1] + x[-2]], range(n-2), [0, 1])
    
    # Menggunakan reduce initializer
    fibonacci_inisial = reduce(lambda x, y: x + [x[-1] + x[-2]], range(n-2), [0, 1])
    
    return fibonacci_biasa, fibonacci_inisial

fibonacci_biasa, fibonacci_inisial = fibonacci(10)

print("Menggunakan reduce biasa =", fibonacci_biasa)
print("Menggunakan reduce initializar =", fibonacci_inisial)
```

Dalam fungsi fibonacci(), terdapat dua implementasi penggunaan reduce(), yaitu menggunakan argumen biasa dan argumen inisialisasi.
Dalam argumen biasa, fungsi reduce() diterapkan pada objek range(n-2) yang menghasilkan bilangan bulat mulai dari 0 hingga n-3 (indeks ke-n-3 dalam deret fibonacci). Setiap elemen dalam range(n-2) digunakan untuk membangun deret fibonacci dengan rumus x[-1] + x[-2] (x[-1] dan x[-2] adalah dua elemen terakhir dalam list x), yang ditambahkan ke dalam list x. List ini dimulai dengan elemen [0, 1]. Akhirnya, list yang dihasilkan dijadikan output dari fungsi reduce().

Dalam argumen inisialisasi, hasil yang dikeluarkan sama dengan argumen biasa, namun pada argumen inisialisasi list dimulai dengan [0, 1] tanpa perlu menghitung elemen dari 0 hingga n-3.

## Buat sebuah program untuk membuat dere fibonacci dari 0 hingga N dengan menggunakan fungsi non-rekursif dan rekursif! Bandingkan keduanya jika nilai N = 500, manakah yang lebih baik? Jelaskan!

```python
def rekursif(n):
  if n <= 1:
    return n
  else:
    return(rekursif(n - 1) + rekursif(n - 2))

N = 20
# Jika menggunakan 500 maka proses running nya kan sangat lama, jadi disini kita hanya menggunkan 20


print("Deret Fibonacci Menggunakan Fungsi Rekursif: ")
for i in range(N + 1):
  print(rekursif(i))
```

```python
def nonrekursif(n):
  if n <= 1:
    return n
  else:
    a, b = 0, 1
    for i in range(2, n + 1):
      c = a + b
      a = b
      b = c
    return b

N = 500

print("Deret Fibonacci Menggunakan Fungsi Non-rekursif: ")
for i in range(N + 1):
  print(nonrekursif(i))
```

Jika nilai N adalah 500, menggunakan implementasi fungsi rekursif akan memakan waktu yang lebih lama dibandingkan dengan menggunakan implementasi fungsi non-rekursif. Hal ini disebabkan karena fungsi rekursif akan memanggil dirinya sendiri berulang-ulang, sehingga setiap nilai deret Fibonacci perlu dihitung ulang di setiap level rekursif, yang memakan waktu lebih lama. Di sisi lain, implementasi fungsi non-rekursif hanya perlu menghitung setiap nilai deret Fibonacci dari 0 hingga N hanya sekali, dan kemudian hanya menggunakan nilai sebelumnya untuk menghitung nilai selanjutnya dalam deret. Hal ini membuat implementasi fungsi non-rekursif lebih cepat dan lebih efisien dalam menghasilkan output.

Dalam kesimpulannya, implementasi fungsi non-rekursif lebih disarankan untuk menghitung deret Fibonacci dengan nilai N yang besar karena lebih cepat dan lebih efisien dalam eksekusinya.
