```python
import pandas as pd
import numpy as np
```

```python
import pandas as pd

# Path ke file CSV Anda
file_path = '/content/dataKP.csv'

# Mengimpor data ke DataFrame
data = pd.read_csv(file_path)

# Menampilkan beberapa baris pertama dari DataFrame
print(data.head())
```

```python
# Membuat DataFrame dari data pegawai
df = pd.DataFrame(data)
```

```python
# Definisikan variabel fuzzy untuk kedisiplinan dan kinerja
kedisiplinan = ctrl.Antecedent(np.arange(0, 101, 1), 'kedisiplinan')
kinerja = ctrl.Antecedent(np.arange(0, 101, 1), 'kinerja')
pegawai_terbaik = ctrl.Consequent(np.arange(0, 101, 1), 'pegawai_terbaik')
```

```python
# Membership functions untuk kedisiplinan
kedisiplinan['cukup'] = fuzz.trapmf(kedisiplinan.universe, [0, 0, 50, 75])
kedisiplinan['sangat baik'] = fuzz.trapmf(kedisiplinan.universe, [75, 100, 100, 100])
```

```python
# Membership functions untuk kinerja
kinerja['rendah'] = fuzz.trapmf(kinerja.universe, [0, 0, 30, 50])
kinerja['sedang'] = fuzz.trapmf(kinerja.universe, [50, 60, 70, 80])
kinerja['tinggi'] = fuzz.trapmf(kinerja.universe, [80, 100, 100, 100])
```

```python
# Membership functions untuk pegawai terbaik
pegawai_terbaik['bukan terbaik'] = fuzz.trapmf(pegawai_terbaik.universe, [0, 0, 30, 70])
pegawai_terbaik['terbaik'] = fuzz.trapmf(pegawai_terbaik.universe, [70, 100, 100, 100])
```

```python
# Aturan-aturan fuzzy
rule1 = ctrl.Rule(kedisiplinan['cukup'] & kinerja['rendah'], pegawai_terbaik['bukan terbaik'])
rule2 = ctrl.Rule(kedisiplinan['cukup'] & kinerja['sedang'], pegawai_terbaik['bukan terbaik'])
rule3 = ctrl.Rule(kedisiplinan['cukup'] & kinerja['tinggi'], pegawai_terbaik['bukan terbaik'])
rule4 = ctrl.Rule(kedisiplinan['sangat baik'] & kinerja['rendah'], pegawai_terbaik['bukan terbaik'])
rule5 = ctrl.Rule(kedisiplinan['sangat baik'] & kinerja['sedang'], pegawai_terbaik['terbaik'])
rule6 = ctrl.Rule(kedisiplinan['sangat baik'] & kinerja['tinggi'], pegawai_terbaik['terbaik'])
```

```python
# Sistem kontrol fuzzy
pegawai_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6])
pegawai_evaluation = ctrl.ControlSystemSimulation(pegawai_ctrl)
```

```python
# Menambahkan kolom hasil fuzzy
df['Skor Pegawai Terbaik'] = 0
```

```python
# Evaluasi setiap pegawai
for index, row in df.iterrows():
    pegawai_evaluation.input['kedisiplinan'] = row['kedisiplinan']
    pegawai_evaluation.input['kinerja'] = row['rata-rata kinerja']
    pegawai_evaluation.compute()
    df.at[index, 'Skor Pegawai Terbaik'] = pegawai_evaluation.output['pegawai_terbaik']
```

```python
# Menampilkan DataFrame dengan hasil evaluasi fuzzy
print(df)
```

```python
# Mengurutkan DataFrame berdasarkan kolom 'Best Employee Score' dari yang tertinggi
df_sorted = df.sort_values(by='Skor Pegawai Terbaik', ascending=False)

# Menampilkan DataFrame yang sudah diurutkan
print(df_sorted)
```

```python
# Menampilkan 3 pegawai dengan nilai komposit tertinggi
top_3_pegawai = df_sorted.head(3)

# Menambahkan kolom 'Pegawai Terbaik' dengan nilai 'Ya' untuk 3 pegawai teratas
top_3_pegawai['Pegawai Terbaik'] = 'Ya'

# Menampilkan kolom yang diperlukan saja
result = top_3_pegawai[['Nama Pegawai', 'kedisiplinan', 'rata-rata kinerja', 'Skor Pegawai Terbaik', 'Pegawai Terbaik']]

# Menampilkan hasil
print(result)
```
