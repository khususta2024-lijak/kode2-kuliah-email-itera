```python
from urllib.request import urlopen
from itertools import groupby
from operator import itemgetter
from ijson import items

url2 = 'https://api.github.com/search/repositories?q=data'
f = urlopen(url2)
repos = items(f, 'items.item')

keyfunc = itemgetter('language')

# Hint: Lakukan filterisasi dengan 'keyfunc' sebagai function dan 'repos' sebagai iterable(s)
cleaned = filter(keyfunc, repos)
# Hint: Lakukan penyortiran dengan iterable(s) menggunakan data yang telah difilter dan 'keyfunc'  sebagai kunci perbandingan
records = sorted
# Hint: Lakukan grouping data menggunakan data yang telah disortir berdasarkan 'keyfunc'
grouped = groupby(records, key = keyfunc)

# Hint: Buatlah sebuah perulangan pada object 'grouped' dengan parameter indeks acuan 'key' dan 'group' sebagai value, yang menghitung total
#       jumlah pengamat ['watchers'] untuk setiap kelompok menggunakan 'sum', dan mencetak 'key' dan total jumlah pengamat (watchers) untuk 
#       setiap kelompok. Kemudian dicetak kedua variabel tersebut menjadi output yang tertera di bawah!
for k, g in grouped:
    total = sum(item['watchers'] for item in g)
    print(k, total)
```

```python
from urllib.request import urlopen
from operator import itemgetter
from functools import partial
from meza import process as pr, fntools as ft
from meza.io import read_json

url4 = 'https://api.github.com/search/repositories?q=data&sort=stars&order=desc'
f = urlopen(url4)
records = read_json(f, path='items.item')

# repos without a language have a value of None, which meza doesn't like
filled = pr.fillempty(records, value='', fields=['language'])
filled, preview = pr.peek(filled)
preview[0]
```

```python
# meza doesn't do well with nested dicts

# Hint: Assign variabel untuk membuat dictionary bersarang dengan perulangan dan fungsi dict juga fntools untuk mengubah dimensi array dalam variabel filled
flat = (dict(ft.flatten(i)) for i in filled)

# Hint: Assign variabel untuk melihat data process
flat, preview = pr.peek(flat)

# Hint: Memanggil variabel preview dengan indeks awal
preview[0]
```

```python
# `watchers` is the pivot field to aggregate by
# `language` is the pivot field to group by
# Hint: Buat variabel universal dengan parameter aggregasi watchers dan di group berdasarkan language
args = ('watchers', 'language')

# the pivot fields we want to include in each row
# Hint: Assign variabel rows dengan field has_pages dan owner_type
rows = ['has_pages', 'owner_type']
# Hint: Assign variabel untuk normalisasi variabel flat dengan parameter "universal"
pivotted = pr.pivot(flat, *args, rows=rows, op=sum)
# Hint: Assign variabel untuk melihat data pivotted
pivotted, preview = pr.peek(pivotted)
# Hint: Memanggil variabel preview dengan indeks awal
preview[0]
```

```python
# Hint: Assign variabel dengan parameter berisi baris dan tidak di normalize (invert)
kwargs = {"rows": rows, "invert": True}
# Hint: Assign variabel untuk normalisasi variabel pivotted dengan dua parameter "universal"
normal = pr.normalize(pivotted, *args , **kwargs)
# Hint: Assign variabel untuk melihat data normal
normal, preview = pr.peek(normal)
# Hint: Memanggil variabel preview dengan indeks awal
preview[0]
```

```python
# Hint: Aggregasi dengan 'watchers
agg_keyfunc = itemgetter('watchers')
# group by `has_pages` and `owner_type`
# Hint: Membuat list tuple menggunakan lambda dengan perulangan dalam 'rows'
group_keyfunc = lambda x: tuple(x[r] for r in rows)
# Hint: Membuat aggregator fungsi baru partial dengan parameter max dan parameter key variabel diawal
aggregator = partial(max, key=agg_keyfunc)
# Hint: Assign variabel universal dengan parameter tidak pakai tuple, dan aggregator
kwargs = {'tupled': False, 'aggregator': aggregator}
# Hint: Pengelompokkan variabel normal dengan parameter universal diatas berdasarkan group_keyfunc
grouped = pr.group(normal, group_keyfunc, **kwargs)
# Hint: Assign variabel untuk melihat data normal
grouped, preview = pr.peek(grouped)
# Hint: Memanggil variabel preview dengan indeks awal
preview[0]
```

```python
from pprint import pprint
# Hint: Assign variabel group untuk mengurutkan data dengan indeks agg_keyfunc, dan diurutkan 
sgrouped = sorted(grouped, key=agg_keyfunc, reverse=True)
# Hint: Membuat perulangan untuk mengembalikan hasil dari dictionary yang telah diurutkan
for group in sgrouped:
    pprint(group)
```
