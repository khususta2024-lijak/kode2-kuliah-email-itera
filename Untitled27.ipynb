```python
import pandas as pd
import csv
with open("/content/budget_data.csv") as csvfile:
  csv_reader = csv.reader (csvfile, delimiter=',') #membaca file csv
  csv_header = next(csv_reader) #membaca header
  print (f"Header: {csv_header}") # memunculkan header

  # inisiasi variabel
  total_months = 0 # menyimpan jumlah total bulan = baris data
  net_total = 0 # menyimpan total laba/rugi dari semua bulan
  previous_profit_loss = None #meyimpan nilai dari bulan sebelumnya (nanti baris 2 dikurang baris 1)
  changes = [] #list yang nyimpan perubahan laba/rugi dari bulan ke bulan
  dates = [] #list yang nyimpan tanggal kalo terjadi perubahan
  greatest_increase = {"amount": float('-inf'), "date": ""} #menyimpan kenaikan laba terbesar dan tanggal
  greatest_decrease = {"amount": float('inf'), "date": ""} #menyimpan penurunan laba terbesar dan tanggal

  for row in csv_reader:
    date = row[0] #kolom pertama disimpan sebagai tanggal
    profit_loss = int(row [1]) #kolom kedua disimpan sebagai laba/rugi
    print(row) #menampilkan kolom

    #count total months
    total_months += 1 #menambah nilai 1 ke total months setiap kali baris diproses

    #count net total
    net_total += profit_loss #menambah nilai ke net_total dari hasil profit+loss

    #calculate changes in "Profit/Losses"
    if previous_profit_loss is not None: #jika laba/rugi ada di bulan sebelumnya maka
      change = profit_loss - previous_profit_loss #mengurangkan baris kedua dan pertama. selisih antara bulan ini dan sebelumnya
      changes.append(change) #menyimpan perubahan
      dates.append(date) #menyimpan perubahan tanggal

      # Check for greatest increase (kenaikan)
      if changes > greatest_increase ["amount"]: #perubahan laba saat ini lebih besar dari peningkatan laba terbesar sebelumnya
        greatest_increase ["amount"] = change #jika iya, perubahan disimpan dalam greatest_increase
        greatest_increase ["date"] = date

      # Check for greatest decrease (penurunan)
      if changes < greatest_decrease ["amount"]: #perubahan laba saat ini lebih kecil dari penurunan laba terbesar sebelumnya
        greatest_decrease ["amount"] = change # jika iya, nilai perubahan disimpan di greatest_decrease
        greatest_decrease ["date"] = date

        # menyimpan laba/rugi bulan ini untuk digunakan pada bulan berikutnya
        previous_profit_loss = profit_loss
```
