```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA,TruncatedSVD
from sklearn.manifold import TSNE
```

### The Digits Dataset

```python
from sklearn.datasets import load_digits
digits = load_digits()
print(digits.DESCR)
```

```python
data = digits.images[0]
```

```python
#Display the first digit
plt.imshow(data, cmap = 'gray', interpolation='nearest')
plt.xticks(np.arange(-0.5, 8.5, 1))
plt.yticks(np.arange(-0.5, 8.5, 1))
plt.grid(True)
plt.show()
```

```python
digits.keys()
```

```python
digits.images[0].shape
```

```python
digits.data[0].shape
```

```python
digitsX = digits.data
digitsY = digits.target
```

### PCA on the Digits Dataset - 2 Componens

```python
pca2 = PCA(n_components = 2)
digits_pca2 = pca2.fit_transform(digitsX)
```

```python
print(pca2.explained_variance_ratio_)
print(pca2.singular_values_)
```

```python
digits_pca2 = pd.DataFrame(data = digits_pca2, columns = ['comp1', 'comp2'])
digits_pca2['label'] = digitsY
```

```python
# Set style of scatterplot
sns.set_context("notebook", font_scale=1.1)
sns.set_style("ticks")

# Create scatterplot of dataframe
sns.lmplot(x='comp1',
           y='comp2',
           data=digits_pca2,
           fit_reg=False,
           legend=True,
           height=9,
           hue = 'label',
           scatter_kws={"s":200, "alpha":0.3})

plt.title('PCA Results: Digits', weight='bold').set_fontsize('14')
plt.xlabel('Principal Comp 1', weight='bold').set_fontsize('10')
plt.ylabel('Principal Comp 2', weight='bold').set_fontsize('10')
```

### SVD on Digits- 2 Components

```python
svd2 = TruncatedSVD(n_components = 2)
```

```python
digits_svd2 = svd2.fit_transform(digitsX)
```

```python
digits_svd2 = pd.DataFrame(data = digits_svd2, columns = ['comp1', 'comp2'])
digits_svd2['label'] = digitsY
```

```python
# Set style of scatterplot
sns.set_context("notebook", font_scale=1.1)
sns.set_style("ticks")

# Create scatterplot of dataframe
sns.lmplot(x='comp1',
           y='comp2',
           data=digits_svd2,
           fit_reg=False,
           legend=True,
           height=9,
           hue = 'label',
           scatter_kws={"s":200, "alpha":0.3})

plt.title('SVD Results: Digits', weight='bold').set_fontsize('14')
plt.xlabel('Feature 1', weight='bold').set_fontsize('10')
plt.ylabel('Feature 2', weight='bold').set_fontsize('10')
```

### TSNE on Digits - 2 Components

```python
tsne2 = TSNE(n_components = 2)
digits_tsne2 = tsne2.fit_transform(digitsX)
```

```python
digits_tsne2 = pd.DataFrame(data = digits_tsne2, columns = ['comp1', 'comp2'])
digits_tsne2['label'] = digitsY
```

```python
# Set style of scatterplot
sns.set_context("notebook", font_scale=1.1)
sns.set_style("ticks")

# Create scatterplot of dataframe
sns.lmplot(x='comp1',
           y='comp2',
           data=digits_tsne2,
           fit_reg=False,
           legend=True,
           height=9,
           hue = 'label',
           scatter_kws={"s":200, "alpha":0.3})

plt.title('TSNE Results: Digits', weight='bold').set_fontsize('14')
plt.xlabel('Principal Comp 1', weight='bold').set_fontsize('10')
plt.ylabel('Principal Comp 2', weight='bold').set_fontsize('10')
```

## Analisis PCA on the Digits Dataset 2 & 3 Componens

### PCA on the Digits Dataset 2 Componens
* Menghasilkan scatter plot 2D dengan
sumbu x dan y mewakili dua komponen utama.
* Lebih mudah diinterpretasikan karena hasilnya berada dalam dua dimensi, sehingga lebih mudah untuk memahami pola dan hubungan antara komponen.
* Dengan menggunakan 2 komponen, kita mendapatkan visualisasi 2D yang memberikan gambaran umum tentang pola dalam data digit.

### PCA on Digits 3 Components
* Menghasilkan scatter plot 3D dengan sumbu x, y, dan z mewakili tiga komponen utama.
* Lebih sulit diinterpretasikan secara visual karena data direduksi ke dalam tiga dimensi. Memahami hubungan antara tiga komponen ini melibatkan pemahaman yang lebih kompleks.
* Dengan 3 komponen, kita mendapatkan visualisasi 3D yang memungkinkan untuk melihat lebih banyak kompleksitas dalam pola data



## Analisis TSNE on the Digits Dataset 2 & 3 Componens

### TSNE on the Digits Dataset 2 Componens
* Mengurangi dimensi dari dataset menjadi 2 komponen utama.
* Menangkap pola dan struktur nonlinear dalam dua dimensi, dengan fokus pada mempertahankan jarak relatif antara sampel.
* Biasanya memerlukan waktu pemrosesan yang lebih singkat karena memproyeksikan data ke dalam dua dimensi.
* Penggunaan 2 komponen menghasilkan visualisasi yang lebih sederhana namun masih dapat memperlihatkan pola umum dalam data.

### TSNE on Digits 3 Components
* Mengurangi dimensi dari dataset menjadi 3 komponen utama
* Menangkap pola dan struktur nonlinear dalam tiga dimensi, mempertahankan jarak relatif antara sampel dengan lebih baik daripada pada 2D, tetapi dengan peningkatan kompleksitas visualisasi.
* Memerlukan waktu pemrosesan yang lebih lama karena kompleksitas pemetaan data ke dalam tiga dimensi.
* Penggunaan 3 komponen memberikan visualisasi yang lebih kompleks dan dapat menangkap lebih banyak variasi dalam data, tetapi memerlukan pemahaman visualisasi tiga dimensi yang lebih mendalam.


## Analisis SVD on Digits 2 & 3 Components

### SVD on Digits 2 Components
* Memperoleh dua komponen utama yang paling signifikan dari data, namun, mungkin kehilangan beberapa kompleksitas dari data asli karena hanya dua komponen yang digunakan.
* menggunakan lebih sedikit komponen memberikan gambaran yang lebih sederhana namun mungkin kehilangan beberapa kompleksitas dalam data.

### SVD on Digits 3 Components
* Memperoleh tiga komponen utama yang paling signifikan dari data. Mempertahankan lebih banyak kompleksitas dari data asli karena menggunakan tiga komponen.
* Penggunaan lebih banyak komponen dapat memberikan gambaran yang lebih komprehensif tentang struktur data, tetapi juga memperumit interpretasi visual.

### Data Transformation (Label Encoding)

```python
import pandas as pd
import numpy as np
dt = "/content/mark.csv"
df = pd.read_csv(dt, header = 0)
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
data_col_category = df.select_dtypes(exclude=[np.number]).columns
data_col_category
```

```python
df[data_col_category].head()
```

```python
# Import the LabelEncoder class
from sklearn.preprocessing import LabelEncoder
# Creating the objek instance
label_encoder = LabelEncoder()
for i in data_col_category:
    df[i] = label_encoder.fit_transform(df[i])
print("Label Encoded Data: ")
df.head()
```

```python
# menyimpan dataset baru
df.to_csv("121450065_c2.csv")
```
