# Learn about recursion

Recursion is essential for functional programming. Learn about it!

Below are a few pratice problems.

Imagine you have a tree of nested lists, like this:

```
[1, [[2], 3], [4], 5, [6, 100, [[7], [[8]], 9]], 10]
```

Write functions to recursively…

calculate the sum of all the numbers in a tree,
calculate the depth of a tree (in the tree above, 8 is the deepest element),
find the largest value in a tree.

You can also try the Towers of Hanoi problem. There are three rods, A, B and C. On rod A there are a stack of 6 disks (with a hole in the middle) of different sizes, sorted by size, so that the biggest disk is on the bottom. Your job is to print out instructions for how to move the stack of disks from rod A to rod C. You can only move one disk at a time, the topmost disk on a rod, and at no point can a bigger disk be on top of a smaller disk. You can of course place disks on rod B temporarily.

So, for example, you can output:

Move disk from A to B

Move disk from A to C

Move disk from B to C

etc.

It is assumed that you move the only disk you can move, the topmost disk. These instructions would mean that you first take the smallest disk from A to B. Then the next smallest from A to C. Then put the smallest, from B, on top of the next smallest, at C. Print out the instructions needed to move the whole stack of 6 disks from A to C.

````python
# Fungsi untuk menghitung jumlah dari semua angka dalam pohon
def sum_tree(tree): #sum_tree untuk memeriksa apakah tree adalah integer atau bukan. Jika integer, maka mengembalikan nilai integer itu sendiri. Jika tidak, maka dilakukan rekursi untuk setiap subtree dalam tree dan mengembalikan jumlah dari nilai-nilai yang dihasilkan.
    if isinstance(tree, int):
        return tree
    else:
        return sum(sum_tree(subtree) for subtree in tree)

# Fungsi untuk menghitung kedalaman dari pohon
def depth_tree(tree): #depth_tree untuk memeriksa apakah tree adalah integer atau bukan. Jika integer, maka mengembalikan nilai 0. Jika tidak, maka dilakukan rekursi untuk setiap subtree dalam tree dan mengembalikan nilai maksimum dari kedalaman subtree ditambah 1.
    if isinstance(tree, int):
        return 0
    else:
        return 1 + max(depth_tree(subtree) for subtree in tree)

# Fungsi untuk menemukan nilai terbesar dalam pohon
def max_tree(tree): #max_tree untuk memeriksa apakah tree adalah integer atau bukan. Jika integer, maka mengembalikan nilai integer itu sendiri. Jika tidak, maka kita dilakukan rekursi untuk setiap subtree dalam tree dan mengembalikan nilai maksimum dari nilai-nilai yang dihasilkan.
    if isinstance(tree, int):
        return tree
    else:
        return max(max_tree(subtree) for subtree in tree)

# Fungsi untuk menyelesaikan masalah Menara Hanoi
def hanoi(n, source, target, auxiliary): #hanoi menggunakan pendekatan rekursif untuk menyelesaikan masalah Menara Hanoi. Pada setiap rekursi, dengan memindahkan n-1 disk dari tongkat awal ke tongkat bantu. Kemudian, memindahkan disk terbesar dari tongkat awal ke tongkat akhir. Terakhir, memindahkan kembali n-1 disk dari tongkat bantu ke tongkat akhir. Rekursi ini diulang sampai semua disk telah dipindahkan dari tongkat awal ke tongkat akhir.
    if n > 0:
        hanoi(n-1, source, auxiliary, target)
        print("Move disk from {} to {}".format(source, target))
        hanoi(n-1, auxiliary, target, source)

# Contoh penggunaan
tree = [1, [[2], 3], [4], 5, [6, 100, [[7], [[8]], 9]], 10]

print("Sum of all numbers in the tree:", sum_tree(tree))
print("Depth of the tree:", depth_tree(tree))
print("Maximum value in the tree:", max_tree(tree))

hanoi(6, 'A', 'C', 'B') #Pemanggilan fungsi hanoi dengan 6 n, penggunaan auxiliary sebagai B dan permulaan dari A serta target C
````

````python
from functools import reduce #fungsi add dan sub dengan menggunakan reduce

def add(*args): #fungsi add, dengan menggunakan reduce untuk menjumlahkan semua argumen yang diberikan
    return reduce(lambda x, y: x + y, args) #Fungsi lambda x, y: x + y yang diberikan ke reduce akan menjumlahkan dua argumen dan mengembalikan hasilnya.
#Fungsi add menerima argumen menggunakan *args, sehingga bisa menerima sejumlah argumen yang tidak terbatas

def sub(*args): #fungsi sub, kita juga menggunakan reduce, namun dengan modifikasi agar bisa mengurangkan semua argumen kecuali argumen pertama dari argumen pertama
    if len(args) == 1: #len(args) untuk mengecek berapa banyak argumen yang diberikan
        return -args[0] #
    else:
        return reduce(lambda x, y: x - y, args[1:], args[0]) #lambda x, y: x - y pada reduce untuk mengurangkan argumen-argumen tersebut #args[1:] untuk menghilangkan argumen pertama dari list argumen

print(add(1, 2, 3))   # Output: 6
print(sub(5, 1, 2))   # Output: 2
````

## compose

Create a function compose that takes 2 functions and does function composition, i.e. compose(double, negate) should return a function that first calls negate and then double on its argument.

````python
def compose(f, g): #fungsi compose() dengan dua argumen f dan g
    def composition(*args, **kwargs): #mendefinisikan fungsi baru bernama composition() yang mengambil argumen tak terbatas menggunakan *args dan **kwargs
        return f(g(*args, **kwargs)) #Fungsi composition() akan mengembalikan hasil pemanggilan fungsi f dengan argumen hasil pemanggilan fungsi g dengan argumen *args dan **kwargs
    return composition 

#Contoh penggunaan fungsi compose()
def double(x): #mendefinisikan fungsi double
    return 2 * x

def negate(x): #mendefenisikan fungsi negate
    return -x

f = compose(double, negate)
result = f(5) #memanggil f(5) yang mengembalikan nilai -10 karena 5 di-negate menjadi -5, dan kemudian hasilnya dikalikan dengan 2 sehingga menghasilkan -10
print(result)
````

## compose with var args

Create a compose function again that can take an arbitrary number of arguments, and that returns a function that can take an arbitrary number of arguments.

```
compose(negate, double, add)(1, 2, 3) => -12
compose(clist, double, sub)(1, 2, 3) => [-8]
```

````python
def compose(*funcs): #Fungsi compose di atas menerima argumen variabel *funcs, yang akan menerima fungsi-fungsi sebagai parameter
    def inner(*args, **kwargs): #fungsi inner, pertama-tama fungsi terakhir di funcs akan dieksekusi dengan argumen yang diberikan
        result = funcs[-1](*args, **kwargs)
        for f in reversed(funcs[:-1]):
            result = f(result)
        return result
    return inner

def negate(x):
    return -x

def double(x):
    return x * 2

def add(x, y, z):
    return x + y + z

def sub(x, y, z):
    return x - y - z

def clist(x):
    return [-x]

result1 = compose(negate, double, add)(1, 2, 3)
result2 = compose(clist, double, sub)(1, 2, 3)

print(result1)  # output: -12
print(result2)  # output: [-8]
````

### zip

Create a function zip that takes an arbitrary number of sequneces, and zips them, i.e. creates a list of lists, where the inner lists consist of the first elements from the given sequences, then the second elements from the given sequences, and so on.

```
zip([1, 2, 3], [4, 5, 6]) => [[1, 4], [2, 5], [3, 6]]
zip([1, 2, 3], [4, 5, 6], [7, 8, 9]) => [[1, 4, 7], [2, 5, 8], [3, 6, 9]]
```

````python
def zip(*args):
    # Menghitung panjang terpendek dari argumen
    length = min(len(seq) for seq in args)
    # Mengembalikan list berisi tuple yang berisi elemen dari setiap argumen
    return [tuple(seq[i] for seq in args) for i in range(length)]

# Contoh penggunaan fungsi zip
a = [1, 2, 3]
b = [4, 5, 6]
c = [7, 8, 9]

print(zip(a, b))  # [(1, 4), (2, 5), (3, 6)]
print(zip(a, b, c))  # [(1, 4, 7), (2, 5, 8), (3, 6, 9)]
````

Create a function zipwith that takes a function f and an arbitrary number of sequences, and returns a list of f applied to the first elements of the given seqences, followed by f applied to the second elements of the sequences, an so on.

In some languages, map can act like this. If it does in your language, you can skip this exercise.

````python
def zipwith(f, *args): #pengambilan 2 argumen 
    return list(map(lambda x: f(*x), zip(*args)))

def add(x, y, z=0): #pengambilan 3 argumen
    return x + y + z

print(zipwith(add, [1, 2, 3], [4, 5, 6]))     # Output: [5, 7, 9]
print(zipwith(add, [1, 2, 3], [4, 5, 6], [1, 1, 1]))    # Output: [6, 8, 10]
````

## car and cdr

cons(a, b) constructs a pair, and car(pair) and cdr(pair) returns the first and last element of that pair. For example, car(cons(3, 4)) returns 3, and cdr(cons(3, 4)) returns 4.

You are given this implementation of cons:

```
def cons(a, b):
    def pair(f):
        return f(a, b)
    return pair
```

Implement car and cdr.

````python
def cons(a, b): #mendefinisikan 2 argumen yang ada
    def pair(f):
        return f(a, b)
    return pair

def car(pair): #mendefinisikan argumen yang ada
    def first(a, b):
        return a
    return pair(first)

def cdr(pair): #mendefinisikan argumen yang ada
    def last(a, b):
        return b
    return pair(last)

p = cons(3, 4)
print(car(p)) # output: 3
print(cdr(p)) # output: 4
````

## flip

Create a function flip that takes a function and flips the first and second argument to it.

```
flip(clist)(1, 2, 3) => [2, 1, 3]
flip(sub)(10, 1) => -9
```

````python
def flip(func):
    def flipped(*args):
        return func(*args[::-1])
    return flipped

def clist(a, b, c):
    return [a, b, c]

def sub(a, b):
    return a - b

flipped_clist = flip(clist)
print(flipped_clist(1, 2, 3)) # Output: [2, 1, 3]

flipped_sub = flip(sub)
print(flipped_sub(10, 1)) # Output: -9
````

## flips

Create a function flips that takes a function and reverses the arguments to it.

```
flips(clist)(1, 2, 3) => [3, 2, 1]
flips(sub)(1, 2, 3) => 0
```

````python
def flips(fn):
    def flipped_fn(*args):
        return fn(*reversed(args))
    return flipped_fn

def clist(a, b, c):
    return [a, b, c]

def sub(a, b, c):
    return a - b - c

flipped_clist = flips(clist)
flipped_sub = flips(sub)

print(flipped_clist(1, 2, 3))  # Output: [3, 2, 1]
print(flipped_sub(1, 2, 3))  # Output: 0
````

## take

Create a function take that takes a number n and a sequence seq, and returns a list of the first n elements of seq.

```
take(3, range(10)) => [0, 1, 2]
```

````python
def take(n, seq):
    result = []
    for i in range(n):
        try:
            result.append(seq[i])
        except IndexError:
            break
    return result

print(take(3, range(10))) # Output: [0, 1, 2]
````

## every_pred

Create a function every_pred that takes an arbitrary number of predicates (functions that return a truth value) and returns a function that returns true if and only if all predicates were truthy for the argument.

```
every_pred(positive, even)(8) => True
every_pred(positive, even)(7) => False
```

````python
def every_pred(*predicates):
    def inner(x):
        return all(p(x) for p in predicates)
    return inner

positive = lambda x: x > 0
even = lambda x: x % 2 == 0

pred = every_pred(positive, even)
print(pred(8)) # True
print(pred(7)) # False
````

frequencies("aabcbcac")
=> {'a': 3, 'c': 3, 'b': 2}
frequencies([1, 2, 2, 2])
=> {1: 1, 2: 3}

## merge_with

Create a function merge_with that takes a function and an arbitrary number of maps, and merges the maps, combining elements that exist in more than 1 using the given function.

```
merge_with(add, {"a": 1, "b": 2}, {"b": 2})
=> {"a": 1, "b": 4}
merge_with(clist, {"a": 1, "b": 2}, {"b": 2})
=> {"a": 1, "b": [2 2]}
```

````python
def merge_with(func, *maps): # Fungsi untuk menggabungkan dua atau lebih dictionary
    result = {}
    for map in maps: # Menggunakan map dengan maps yaitu, variabel tuple yang berisi dictionary-dictionary yang akan digabungkan
        for key, value in map.items():
            if key in result:
                result[key] = func(result[key], value)
            else:
                result[key] = value
    return result

def add(x, y):
    return x + y # Mengembalikan nilai yang akan menghasilkan output penggabungan dari 2 dictionaries 

def clist(x, y):
    return [x, y]

print(merge_with(add, {"a": 1, "b": 2}, {"b": 2})) # Output: {'a': 1, 'b': 4}
print(merge_with(clist, {"a": 1, "b": 2}, {"b": 2})) # Output: {'a': 1, 'b': [2, 2]}
````

## tree_seq

Create a function tree_seq that takes a functon is_branch, a function children and a tree t, that returns a list of the nodes of the tree, in depth-first order. is_branch should take 1 argument and return true for nodes that can have children. children should take 1 argument and return a sequence of the children for nodes that is_branch returns true for. t is the root of the tree. Look carefully at the examples below. As far as this function is concerned, a tree is not necessarily a list of lists; a tree is whatever the given functions define it to be.

Assume that is_list and is_integer below tests if the argument is a list or integer, respectively, and that identity is the identity function. range takes an integer n an returns a sequence from 0 to n-1 (this function exists in most languages).

```
t = [1, [2, [3, 5]]]
tree_seq(is_list, identity, t)
=> [[1, [2, [3, 5]]], 1, [2, [3, 5]], 2, [3, 5], 3, 5]
tree_seq(is_integer, range, 3)
=> [3, 0, 1, 0, 2, 0, 1, 0]
```

````python
def tree_seq(is_branch, children, t):
    # Jika node adalah leaf, maka kembalikan node tersebut
    if not is_branch(t):
        return [t]
    # Jika node dapat memiliki child, maka kembalikan node dan child-nya secara rekursif
    return [t] + sum([tree_seq(is_branch, children, c) for c in children(t)], [])

# Fungsi untuk mengecek apakah sebuah data adalah list
def is_list(x):
    return isinstance(x, list)

# Fungsi untuk mengecek apakah sebuah data adalah integer
def is_integer(x):
    return isinstance(x, int)

# Fungsi identitas
def identity(x):
    return x

t = [1, [2, [3, 5]]]
print(tree_seq(is_list, identity, t))
# Output: [[1, [2, [3, 5]]], 1, [2, [3, 5]], 2, [3, 5], 3, 5]

print(tree_seq(is_integer, range, 3))
# Output: [3, 0, 1, 0, 2, 0, 1, 0]
````

## memoize

Create a function memoize take takes a function and memoizes it, i.e. it returns a function that does the same thing that the given function, but caches its results in a map.

```
new_add = memoize(add)
new_add(1, 2, 3) => 6
new_add(1, 2, 3) => 6 (Did not actually compute add(1, 2, 3).)
```

The second call of new_add should not compute using add again, but instead look up the stored result.

````python
# Fungsi memoize
def memoize(f):
    cache = {}
    def new_function(*args):
        if args not in cache:
            cache[args] = f(*args)
        return cache[args]
    return new_function

def add(x, y, z):
    return x + y + z

# Digunakan sebagai fungsi utk penambahan angka
new_add = memoize(add)

new_add(1, 2, 3)  # input angka 1, 2, 3

new_add(1, 2, 3)  # input angka 1, 2, 3
````

## group_by

Create a function group_by that takes a function and a sequence and groups the elements of the sequence based on the result of the given function.

len returns the length of a sequence.

```
group_by(len, ["hi", "dog", "me", "bad", "good"])
=> {2: ["hi", "me"], 3: ["dog", "bad"], 4: ["good"]}
```

````python
# Fungsi Group_by
def group_by(func, seq):
    result = {}
    for item in seq:
        key = func(item)
        if key not in result:
            result[key] = []
        result[key].append(item)
        # Fungsi return untuk pengeluaran output
    return result

# Berikut contoh penggunaannya
group_by(len, ["hi", "dog", "me", "bad", "good"])
````

Create a function update that takes a map m, a key k, a function f and an arbitrary number of additional arguments args. Returns a new map, that is like m, but has the value for k replaced by the result of applying f to the previous value and the given args. If the previous value does not exist, create a new entry for k.

```
bob = {"name": "bob", "hp": 3}
update(bob, "hp", add, 2)
=> {"name": "bob", "hp": 5}
nohp = {"name": "bob"}
update(nohp, "hp", add, 2)
=> {"name": "bob", "hp": 2}
```

````python
def update(m, k, f, *args): # Mendefinisikan fungsi update dengan parameter m, k, f, dan *args
    new_m = m.copy() # Membuat salinan dictionary m menggunakan metode copy()
    if k in new_m: # Memeriksa apakah kunci k sudah ada dalam dictionary baru new_m
        new_m[k] = f(new_m[k], *args) # Jika kunci k sudah ada, maka nilai yang terkait dengan kunci tersebut akan diperbarui menggunakan fungsi f dengan argumen nilai saat ini dan argumen tambahan *args
    else: # Jika kunci k belum ada dalam dictionary baru,
        new_m[k] = f(*args) # maka kunci dan nilai baru akan ditambahkan ke dalam dictionary menggunakan fungsi f dengan argumen tambahan *args
    return new_m # Mengembalikan dictionary new_m yang telah diperbarui

bob = {"name": "bob", "hp": 3} # Mendefinisikan dictionary bob dengan dua pasangan kunci-nilai
print(update(bob, "hp", lambda x, y: x + y, 2)) # Memanggil fungsi update dengan argumen bob sebagai dictionary yang akan diperbarui, hp sebagai kunci yang akan diperbarui, lambda fungsi yang akan menjumlahkan nilai saat ini dengan argumen tambahan, dan 2 sebagai argumen tambahan, lalu dicetak

nohp = {"name": "bob"} # Mendefinisikan dictionary nohp dengan satu pasangan kunci-nilai
print(update(nohp, "hp", lambda x, y: x + y, 2, 0)) # Memanggil fungsi update dengan argumen nohp sebagai dictionary yang akan diperbarui, hp sebagai kunci yang akan diperbarui, lambda fungsi yang akan menjumlahkan nilai saat ini dengan argumen tambahan, dan 2 dan 0 sebagai argumen tambahan, lalu dicetak
````

## update_in

Create a function update_in that is similar to update, but can update a value inside nested maps.

```
a = {"a": 1, "b": {"c": 2, "d": {"e": 3}}}
update_in(a, ["b", "d", "e"], add, 10)
=> {"a": 1, "b": {"c": 2, "d": {"e": 13}}}
update_in(a, ["b", "d", "f"], add, 10)
=> {"a": 1, "b": {"c": 2, "d": {"e": 3, "f": 10}}}
```

````python
def add(a, b): # menerima dua argumen a dan b
    return a + b # mengembalikan hasil penjumlahan dari argumen a dan b

def update_in(m, ks, f, *args):# Mendefinisikan fungsi update dengan parameter m, k, f, dan *args
    if len(ks) == 1: # memeriksa apakah ks hanya memiliki satu key saja
        k = ks[0] # mengambil key pertama dari ks
        if k in m: # memeriksa apakah key tersebut terdapat pada dictionary "m"
            m[k] = f(m[k], *args)
        else:
            m[k] = f(*args)
    else:
        k = ks[0]
        if k in m:
            update_in(m[k], ks[1:], f, *args)
        else:
            m[k] = {} # membuat key baru pada dictionary "m" dengan {}
            update_in(m[k], ks[1:], f, *args) 
    return m # mengembalikan dictionary pada "m"

a = {"a": 1, "b": {"c": 2, "d": {"e": 3}}}
print(update_in(a, ["b", "d", "e"], add, 10))  # Output: {"a": 1, "b": {"c": 2, "d": {"e": 13}}}
print(update_in(a, ["b", "d", "f"], add, 10, 0))  # Output: {"a": 1, "b": {"c": 2, "d": {"e": 3, "f": 10}}}
````

## balanced

Create a function balanced that takes a string s and returns True if it contains only balanced parens, brackets and braces ((), [], {}), otherwise False.

```
balanced("abc(def{g}hi[jk]((()))l)m") => True
balanced("a(b") => False
balanced("([)]") => False
```

````python
def balanced(s): #Mendefinisikan balanced dengan parameter s 
    stack = [] #Membuat stack kosong yang digunakan untuk memeriksa keseimbangan pada tanda kurung
    pairs = {'(': ')', '[': ']', '{': '}'} #Membuat sebuah dictionary yang menyimpan pasangan tanda kurung, yaitu ( : ), [ : ], serta { : }
    for c in s: #Melakukan iterasi pada setiap karakter c dalam string s
        if c in pairs: #Memeriksa apakah suatu nilai c terdapat dalam sebuah struktur data pasangan (pairs) 
            stack.append(c) #Menambahkan elemen c ke dalam tumpukan (stack) yang ada
        elif c in pairs.values(): #Memeriksa apakah nilai c terdapat di dalam nilai-nilai dari dictionary pairs
            if not stack or pairs[stack.pop()] != c: #Memeriksa apakah stack kosong atau apakah elemen teratas dalam stack tidak sesuai dengan karakter c
                return False #Mengembalikan False
    return not stack #Mengembalikan True jika stack kosong

#Memanggil fungsi "balanced" dengan beberapa contoh string untuk menguji keseimbangan tanda kurung, kemudian dicetak untuk menghasilkan output
print(balanced("abc(def{g}hi[jk]((()))l)m"))
print(balanced("a(b"))
print(balanced("([)]"))
````

## postwalk, prewalk

Create two functions postwalk and prewalk, that both take a function f and a tree t. (A tree in this case is a list where the elements can be trees themselves), e.g. [1, [2, 3, [1, 2]], [4, 5, 6]]. In this case 1, [2, 3, [1, 2] and 6 are all elements of the tree (but not the only ones).) The functions should apply f to each element in the tree to create a new tree with the results, like map but for trees instead of lists. The difference between prewalk and postwalk is that prewalk does the replacement on the way “down” and postwalk on the way “up”, so that prewalk can continue down into the new value that was “just” created.

In the following example, imagine that wrap_unless_zero, if the argument is a number, wraps the argument in a list if it is larger than zero, then decrements it by 1.

```
postwalk(wrap_unless_zero, [1 [2 [3]]]) => [[0] [[1] [[2]]]]
prewalk(wrap_unless_zero, [1 [2 [3]]]) => [[0] [[[0]] [[[[0]]]]]]
```

````python
def postwalk(f, tree): # Mendefinisikan fungsi postwalk dengan parameter f dan tree
    if isinstance(tree, list): # Jika elemen merupakan sebuah list, 
        return [postwalk(f, x) for x in tree] # maka penelusuran pasca-order dilakukan secara rekursif pada setiap elemen list tersebut
    else: # Jika elemen bukan sebuah list,
        return f(tree) # maka fungsi f diterapkan pada elemen tersebut

def prewalk(f, tree): # Mendefinisikan fungsi prewalk dengan parameter f dan tree
    if isinstance(tree, list): # Jika elemen merupakan sebuah list,
        return f([prewalk(f, x) for x in tree]) # maka penelusuran pra-order dilakukan secara rekursif pada setiap elemen list tersebut
    else: # Jika elemen bukan sebuah list,
        return f(tree) # maka fungsi f diterapkan pada elemen tersebut

def wrap_unless_zero(x): # Mendefinisikan fungsi wrap_unless_zero dengan parameter x
    if isinstance(x, int): # Jika elemen x adalah integer, maka dilakukan pengecekan kondisi
        if x > 0: # Jika x lebih besar dari 0,
            return [x - 1] # maka dikembalikan list dengan elemen x - 1
        elif x == 0: # Jika x sama dengan 0, 
            return [0] # maka dikembalikan list dengan elemen 0
        else: # Jika x negatif, 
            return x # maka dikembalikan nilai x tanpa perubahan
    else: # Jika elemen x bukan integer,
        return x # maka dikembalikan nilai x tanpa perubahan

print(postwalk(wrap_unless_zero, [1, [2, [3]]])) #Memanggil fungsi postwalk dengan parameter wrap_unless_zero sebagai fungsi transformasi dan struktur data pohon [1, [2, [3]]], lalu dicetak
print(prewalk(wrap_unless_zero, [1, [2, [3]]])) #Memanggil fungsi prewalk dengan parameter wrap_unless_zero sebagai fungsi transformasi dan struktur data pohon [1, [2, [3]]], lalu dicetak
````

## Recreate map, filter and reduce

Implement map, filter and reduce from scratch.

````python
def square(x): # Mendefinisikan fungsi square() 
    return x ** 2

numbers = [1, 2, 3, 4, 5] # Membuat sebuah list bernama numbers yang berisi angka-angka
squared_numbers = list(map(square, numbers)) # Menggunakan fungsi map() untuk menerapkan fungsi square() pada setiap elemen dalam list numbers, lalu mengkonversi hasil map() menjadi list menggunakan fungsi list().
print(squared_numbers) # Mencetak hasil list squared_numbers

def is_even(x): # Mendefinisikan fungsi is_even()
    return x % 2 == 0

numbers = [1, 2, 3, 4, 5] # Membuat sebuah list bernama numbers yang berisi angka-angka
even_numbers = list(filter(is_even, numbers)) # Menggunakan fungsi filter() untuk memfilter elemen-elemen dalam list numbers yang memenuhi kondisi yang diberikan oleh fungsi is_even(), lalu mengkonversi hasil filter() menjadi list menggunakan fungsi list()
print(even_numbers) # Mencetak hasil list even_numbers

from functools import reduce # Mengimpor fungsi reduce dari modul functools

def multiply(x, y): # Mendefinisikan fungsi multiply()
    return x * y

numbers = [1, 2, 3, 4, 5] # Membuat sebuah list bernama numbers yang berisi angka-angka
product = reduce(multiply, numbers) # Menggunakan fungsi reduce() untuk mengaplikasikan fungsi multiply() secara berulang pada semua elemen dalam list numbers sehingga menghasilkan nilai produk dari semua elemen tersebut
print(product) # Mencetak nilai product
````
