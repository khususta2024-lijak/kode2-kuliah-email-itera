```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

```

```python
# Baca dataset
data = pd.read_csv('/content/4fc470c4-f6dc-4823-ab08-33a50e570838.csv')
df = data.drop(['_id', 'periode_data', 'stasiun', 'parameter_pencemar_kritis', 'kategori'], axis=1)
df = df[~df.isin(['-', '---']).any(axis=1)]
df
```

```python
# Ubah kolom 'tanggal' menjadi datetime dan urutkan data
data['tanggal'] = pd.to_datetime(data['tanggal'])
data = data.sort_values(by='tanggal')
```

```python
# Pilih kolom yang relevan untuk prediksi, misalnya 'pm_sepuluh'
data_for_prediction = data[['tanggal', 'pm_sepuluh']]  # Anda bisa mengganti kolom sesuai dengan yang ingin diprediksi
```

```python
# Gantilah nilai yang tidak valid (misalnya tanda minus) dengan NaN
data_for_prediction['pm_sepuluh'] = pd.to_numeric(data_for_prediction['pm_sepuluh'], errors='coerce')

# Jika ada nilai NaN, Anda bisa mengisi dengan nilai rata-rata atau menghapus baris tersebut
data_for_prediction['pm_sepuluh'].fillna(data_for_prediction['pm_sepuluh'].mean(), inplace=True)

# Normalisasi data
scaler = MinMaxScaler()
data_for_prediction['Scaled_pm_sepuluh'] = scaler.fit_transform(data_for_prediction[['pm_sepuluh']])

# Cek hasilnya
print(data_for_prediction.head())
```

```python
# Siapkan data untuk RNN
def create_sequences(data, seq_length=6):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i + seq_length])
        y.append(data[i + seq_length])
    return np.array(X), np.array(y)

sequence_length = 6  # Panjang sequence (jumlah hari sebelumnya yang digunakan untuk memprediksi)
X, y = create_sequences(data_for_prediction['Scaled_pm_sepuluh'].values, sequence_length)

# Ubah bentuk X agar sesuai dengan input RNN (samples, timesteps, features)
X = X.reshape((X.shape[0], X.shape[1], 1))
```

```python
# Bangun model RNN dengan lebih banyak layer
model = Sequential([
    SimpleRNN(100, activation='relu', return_sequences=True, input_shape=(sequence_length, 1)),
    SimpleRNN(50, activation='relu'),
    Dense(1)
])
model.compile(optimizer='adam', loss='mse')

# Latih model dengan epoch yang lebih banyak
model.fit(X, y, epochs=200, verbose=0)
```

```python
# Prediksi nilai berikutnya
last_sequence = data_for_prediction['Scaled_pm_sepuluh'].values[-sequence_length:]  # Data 6 hari terakhir
last_sequence = last_sequence.reshape((1, sequence_length, 1))  # Sesuaikan bentuk input

next_day_prediction_scaled = model.predict(last_sequence)
next_day_prediction = scaler.inverse_transform(next_day_prediction_scaled)  # Kembalikan ke skala asli

# Tampilkan hasil prediksi dalam angka
print("Prediksi PM10 untuk hari berikutnya:", next_day_prediction[0, 0])
```

```python
# Data prediksi untuk plot
predicted_values = model.predict(X)
predicted_values_rescaled = scaler.inverse_transform(predicted_values)

# Plot data aktual dan prediksi
plt.figure(figsize=(10, 6))
plt.plot(data_for_prediction['tanggal'][sequence_length:], data_for_prediction['pm_sepuluh'][sequence_length:], label='Data Aktual')
plt.plot(data_for_prediction['tanggal'][sequence_length:], predicted_values_rescaled, label='Prediksi RNN')
plt.xlabel('Tanggal')
plt.ylabel('Polusi')
plt.title('Prediksi Polusi Menggunakan LSTM')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Hitung nilai MSE dan MAE
mse = mean_squared_error(data_for_prediction['pm_sepuluh'][sequence_length:], predicted_values_rescaled)
mae = mean_absolute_error(data_for_prediction['pm_sepuluh'][sequence_length:], predicted_values_rescaled)

# Hitung MAPE
actual = data_for_prediction['pm_sepuluh'][sequence_length:].values
predicted = predicted_values_rescaled[:, 0]
mape = np.mean(np.abs((actual - predicted) / actual)) * 100

print(f"Mean Squared Error (MSE): {mse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
```
