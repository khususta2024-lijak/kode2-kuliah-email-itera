Praktikum 2 : Persiapan Data (Data Preparation)

```python
import pandas as pd
import seaborn as sns
```

```python
# membaca file
df = pd.read_csv("/content/diabetes_RS2.csv")
```

```python
df.head()
```

```python
# checking the number of columns
df.columns
```

```python
# rename the column "No_Pation" to "Patients_No"
df.rename(columns={"No_Pation" : "Patients_No"}, inplace=True)
```

```python
# confirm the change
df.columns
```

```python
# to check for missing values
df.isnull().sum()
```

```python
# replace the missing values in the column "Hb41c"
mean_value = df["HbA1c"].mean()
mean_value
```

```python
# using the mean value to replace the missing records
df["HbA1c"].fillna(mean_value, inplace=True)

# to confirm the change
df.isnull().sum()
```

```python
# dropping the missing values of other columns
df1 = df.dropna()
df1.isnull().sum()
```

```python
#check for information concerning the dataset
df1.info()
```

```python
df1.groupby('CLASS')['CLASS'].agg("count")
```

```python
# check for the unique values in the column "CLASS"
df['CLASS'].unique()
```

```python
# to clean the column, i replace the unique values
df1["CLASS"] = df1["CLASS"].str.replace("Y","Y")
df1["CLASS"] = df1["CLASS"].str.replace("N","N")
```

```python
# to confirm the change
df1["CLASS"].unique()
```

```python
import numpy as np
# variabel Pregnancies
#q1, q3, batas bawah, batas atas, selisih q3 dengan q1
q11,q31 = np.percentile(df['Pregnancies'],[25,75])
s1=q31-q11
ba1=q31+(1.5*s1)
bw1=q11-(1.5*s1)
print(q11)
print(q31)
print(s1)
print(ba1)
print(bw1)
```

```python
dt1=df[(df['Pregnancies']<bw1) | (df['Pregnancies']>ba1)]
dt1
```

```python
#manghapus outlier
data2=df.drop([88,159,298,455])
```

```python
data2.info()
```

```python
#identifikasi data imbalance
data2['Outcome'].unique() #menampilkan variabel pada atribut Outcome
```

```python
data2['Outcome'].value_counts() #menampilkan nilai komposisi data pada atribut Outcome
```

```python
data3 = (data2.groupby('Outcome', as_index=False)
                .apply(lambda x: x.sample(n=252))
                .reset_index(drop=True)) #ambil 252
```

```python
data3.info()
```
