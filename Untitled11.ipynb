```python
import random

def generate_random_value(x_min, x_max):
    return x_min + random.random() * (x_max - x_min)

def generate_random_values():
    x1_min, x1_max = -5, 9.8
    x2_min, x2_max = 0, 7.3

    random_values = []
    for _ in range(10):
        x1 = generate_random_value(x1_min, x1_max)
        x2 = generate_random_value(x2_min, x2_max)
        random_values.append((x1, x2))

    return random_values

# Generate 10 random values
random_values = generate_random_values()

# Print the random values
for i, (x1, x2) in enumerate(random_values, start=1):
    print(f"Nilai ke-{i}: x1 = {x1}, x2 = {x2}")
```

```python
import numpy as np
from pyabc import ABC, Hive, parallel

# Define the objective function
def objective_function(x):
    x1, x2 = x
    return 19 + x1 * np.sin(x1 * np.pi) + (10 - x2) * np.sin(x2 * np.pi)

# Define the bounds
lower_bounds = [-5, 0]
upper_bounds = [9.8, 7.3]
bounds = (lower_bounds, upper_bounds)

# Define the ABC algorithm
abc = ABC(
    Hive(
        n_workers=4,
        memory=True,
        parallel=parallel.ThreadingBackend(),
    ),
    objective_function,
    bounds=bounds,
    minimize=False,
)

# Run the optimization
history = abc.run(max_iterations=100)

# Get the best solution
best_solution = history.best()
print("Best Solution:", best_solution)
```
