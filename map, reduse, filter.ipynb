```python
nilai = [5,6,7,8,9]

uts = lambda x : "lolos" if x>5 else "gagal"
uts(4)
```

```python
penjualan = [('dongren', 90), ('renren', 34), ('jaemren', 98)]

penjualan.sort(key=lambda x : x[1])
# 1 untuk angka

print(penjualan)
```

```python
penjualan = [('dongren', 90), ('renren', 34), ('jaemren', 98)]

penjualan.sort(key=lambda x : x[0])
# 0 untuk huruf

print(penjualan)
```

# MAP

```python
nilai = [1,2,3,4,5,6]
        
hasilkuadrat = list(map(lambda x : x**3, nilai))
# map berfungsi memasukkan fungsi lambda ke variabel nilai

print(hasilkuadrat)
```

```python
 nama = ['KEVIN', 'JOSUA', 'DEDE']

 huruf = list(map(str.lower, nama))
# kebalikan lower itu upper
 print(huruf)
```

# filter

```python
# filter dengan looping

nilai = [20,30,48,50,68,29,40,58,38]
hasil_filter = []

for i in nilai:
  if i > 40:
    hasil_filter.append(i)

print(hasil_filter)
```

```python
nilai = [20,30,48,50,68,29,40,58,38]

def uts(nilai):
  hasil_filter = []
  for i in nilai:
    if i > 40:
      hasil_filter.append(i)
  return hasil_filter

uts(nilai)
```

```python
nilai = [20,30,48,50,68,29,40,58,38]

uts = list(filter(lambda nilai : nilai > 40 , nilai))

print(uts)
```

```python
nama = ['lia', 'liza', 'wulan', 'meliza', 'mey']

panggilanku = list(map(lambda nama : 'nama kesayangan = ()'.format(nama), 
                       filter(lambda nama : len(nama) > 3, nama)))
                       
                     

print(panggilanku)
```

# Reduce

```python
from functools import reduce
nilai = [1,2,3,4,5]

nilaihasil = reduce(lambda x, y : x+y , nilai)
# jadi x=1 y=2 xy=3
# x=3 y=3 xy=6
# x=6 y=4 xy=10
# x=10 y=5 xy=15                  
print(nilaihasil)
```
