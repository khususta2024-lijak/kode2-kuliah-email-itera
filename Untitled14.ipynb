```python
import random

# Batasan untuk x1
x1_min = -5.0
x1_max = 9.8

# Batasan untuk x2
x2_min = 0.0
x2_max = 7.3

# Menginisiasi x1 secara acak
x1 = x1_min + random.random() * (x1_max - x1_min)

# Menginisiasi x2 secara acak
x2 = x2_min + random.random() * (x2_max - x2_min)

print("Nilai x1 yang diinisiasi secara acak:", x1)
print("Nilai x2 yang diinisiasi secara acak:", x2)
```
