```python
import numpy as np

# Konstanta yang digunakan
n = 29
pi = np.pi

# Bagian pertama dari persamaan
factor1 = 2 * 0.5312
sin1 = np.sin((n * 2 * pi * 0.5312))
divisor1 = (n * 2 * pi * 0.5312)
term1 = sin1 / divisor1

# Bagian kedua dari persamaan
factor2 = 2 * 0.2187
sin2 = np.sin((n * 2 * pi * 0.2187))
divisor2 = (n * 2 * pi * 0.2187)
term2 = sin2 / divisor2

# Bagian ketiga (dalam kurung besar) dari persamaan
cos_term = np.cos((2 * pi * n) / 53)
inner_bracket = 0.54 + 0.46 * cos_term

# Menghitung nilai akhir H(29)
H_29 = factor1 * term1 - factor2 * term2 * inner_bracket

# Menampilkan hasil
print(f"H(29) = {H_29:.4f}")
```

```python
import numpy as np

# Konstanta yang digunakan
pi = np.pi
n_values = np.arange(1, 53)  # n dari 1 hingga 52

# List untuk menyimpan hasil H(n)
H_values = []

# Menghitung H(n) untuk setiap n
for n in n_values:
    # Bagian pertama dari persamaan
    factor1 = 2 * 0.5312
    sin1 = np.sin((n * 2 * pi * 0.5312))
    divisor1 = (n * 2 * pi * 0.5312)
    term1 = sin1 / divisor1

    # Bagian kedua dari persamaan
    factor2 = 2 * 0.2187
    sin2 = np.sin((n * 2 * pi * 0.2187))
    divisor2 = (n * 2 * pi * 0.2187)
    term2 = sin2 / divisor2

    # Bagian ketiga (dalam kurung besar) dari persamaan
    cos_term = np.cos((2 * pi * n) / 53)
    inner_bracket = 0.54 + 0.46 * cos_term

    # Menghitung nilai H(n)
    H_n = factor1 * term1 - factor2 * term2 * inner_bracket

    # Menyimpan hasil ke dalam list
    H_values.append(H_n)

# Menampilkan hasil
for n, H_n in zip(n_values, H_values):
    print(f"H({n}) = {H_n:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import get_window, convolve

# 1. Membuat sinyal diskrit
fs = 1000  # Sampling frequency (Hz)
t = np.linspace(0, 1, fs, endpoint=False)  # Time vector 1 second
freq_low = 50   # Low frequency (Hz)
freq_high = 200  # High frequency (Hz)

# Kombinasi sinyal frekuensi rendah dan tinggi
signal = np.sin(2 * np.pi * freq_low * t) + 0.5 * np.sin(2 * np.pi * freq_high * t)

# 2. Membuat Band-pass filter dengan Hamming window
fc1, fc2 = 40, 100  # Cut-off frequencies (Hz)
order = 101  # Filter order
nyquist = fs / 2  # Nyquist frequency

# Normalisasi cut-off frequencies
fc1_norm = fc1 / nyquist
fc2_norm = fc2 / nyquist

# Membuat kernel filter (Hamming window)
ideal_bandpass = np.sinc(2 * fc2_norm * (np.arange(order) - (order - 1) / 2)) - \
                 np.sinc(2 * fc1_norm * (np.arange(order) - (order - 1) / 2))
hamming_window = get_window('hamming', order)
bandpass_filter = ideal_bandpass * hamming_window
bandpass_filter /= np.sum(bandpass_filter)  # Normalisasi amplitudo

# 3. Konvolusi antara sinyal dan filter
filtered_signal = convolve(signal, bandpass_filter, mode='same')

# 4. Ekstraksi fitur menggunakan FFT
fft_signal = np.fft.fft(filtered_signal)
freqs = np.fft.fftfreq(len(filtered_signal), 1 / fs)

# Hanya ambil frekuensi positif
fft_magnitude = np.abs(fft_signal[:len(filtered_signal) // 2])
freqs_positive = freqs[:len(filtered_signal) // 2]

# Plot hasil
plt.figure(figsize=(12, 10))

# Plot sinyal asli
plt.subplot(3, 1, 1)
plt.plot(t, signal)
plt.title("Sinyal Diskrit")
plt.xlabel("Waktu (detik)")
plt.ylabel("Amplitudo")

# Plot sinyal hasil filtering
plt.subplot(3, 1, 2)
plt.plot(t, filtered_signal, color='orange')
plt.title("Sinyal Setelah Filtering (Band-pass)")
plt.xlabel("Waktu (detik)")
plt.ylabel("Amplitudo")

# Plot spektrum hasil FFT
plt.subplot(3, 1, 3)
plt.plot(freqs_positive, fft_magnitude, color='green')
plt.title("Spektrum Frekuensi (FFT)")
plt.xlabel("Frekuensi (Hz)")
plt.ylabel("Magnitudo")

# Visualisasi Impuls Respon
plt.figure(figsize=(8, 4))
plt.stem(bandpass_filter, use_line_collection=True, basefmt=" ")
plt.title("Impuls Respon Filter Band-Pass (Hamming Window)")
plt.xlabel("Sample Index")
plt.ylabel("Amplitudo")
plt.grid()
plt.tight_layout()
plt.show()

plt.tight_layout()
plt.show()
```
