## **2. Identifikasi duplikat data.**

```python
df2.info()
```

```python
df2.groupby('CLASS')['CLASS'].agg("count")
```

```python
df2['CLASS'].unique() # periksa nilai unik di kolom CLASS
```

```python
# untuk membersihkan kolom, maka ganti nilai unik
df2["CLASS"] = df2["CLASS"].str.replace("Y","Y")
df2["CLASS"] = df2["CLASS"].str.replace("N","N")
```

```python
# untuk mengkonfirmasi perubahan
df2["CLASS"].unique()
```

mengecek apakah ada duplikat data pada kolom CLASS dengan menggunakan groupby. kemudian ditemukan adanya duplikat data yaitu pada N dan Y. lalu dilakukannya replace sehingga tidak ada lagi duplikat data
