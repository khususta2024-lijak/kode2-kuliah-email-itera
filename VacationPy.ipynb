### Step 1: Create a map that displays a point for every city in the `city_data_df` DataFrame. The size of the point should be the humidity in each city.

```python
import folium #membuat peta
import pandas as pd
from IPython.display import IFrame, display #menampilkan peta secara langsung


# Create a map centered around the average coordinates
map_center = [city_data_df['Lat'].mean(), city_data_df['Lng'].mean()] #menghitung pusat peta dengan menggunakan rata-rata dari kolom Latitude dan Longitude
city_map = folium.Map(location=map_center, zoom_start=2) #membuat objek peta dengan lokasi pusat yang ditentukan dan zoom level 2 untuk memberikan tampilan global

# Define a color palette based on humidity levels
def get_color(humidity):
    if humidity >= 90: #Kelembaban 90% ke atas.
        return 'red'
    elif humidity >= 80: #80% hingga 89%.
        return 'orange'
    elif humidity >= 70: #70% hingga 79%.
        return 'yellow'
    else:
        return 'green' #Di bawah 70%.

# Add a point for each city in the DataFrame
for index, row in city_data_df.iterrows(): #Iterasi setiap baris dalam DataFrame yang berisi informasi kota
    color = get_color(row['Humidity'])  # Get color based on humidity
    folium.CircleMarker(
        location=[row['Lat'], row['Lng']], #Menambahkan lingkaran di lokasi yang sesuai (Lat, Lng)
        radius=row['Humidity'] / 10,  # ukuran yang proporsional terhadap kelembaban (Humidity/10)
        popup=( #Informasi tambahan yang ditampilkan ketika marker di-klik
            f"Lat: {row['Lat']}<br>"
            f"Lng: {row['Lng']}<br>"
            f"City: {row['City']}<br>"
            f"Country: {row['Country']}<br>"
            f"Humidity: {row['Humidity']}%"
        ),
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.6
    ).add_to(city_map)

# Add a legend to the map
legend_html = """
    <div style="position: fixed;
                top: 10px; left: 10px; width: 150px; height: auto;
                border:2px solid grey; z-index:9999; font-size:14px;
                background-color:white;">
        <p style="text-align:center;">Humidity Legend</p>
        <p style="color:red;">&bull; 90% and above (High)</p>
        <p style="color:orange;">&bull; 80% - 89% (Moderate)</p>
        <p style="color:yellow;">&bull; 70% - 79% (Low)</p>
        <p style="color:green;">&bull; Below 70% (Very Low)</p>
    </div>
"""
city_map.get_root().html.add_child(folium.Element(legend_html)) #Menambahkan HTML legend ini ke peta sehingga tampil sebagai informasi tambahan

# Save the map to an HTML file
city_map.save("city_humidity_map2.html")

# Display the map in Google Colab
city_map
```

### Step 2: Narrow down the `city_data_df` DataFrame to find your ideal weather condition

```python
# Step 2: Narrow down the city_data_df DataFrame to find your ideal weather condition

# Define your ideal weather conditions
max_temp_min = 18  # Minimum temperature (C)
max_temp_max = 35  # Maximum temperature (C)
max_humidity = 100   # Maximum humidity (%)
max_cloudiness = 0  # Maximum cloudiness (%)
max_wind_speed = 5  # Maximum wind speed (m/s)

# Narrow down cities that fit criteria and drop any results with null values
filtered_cities = city_data_df[
    (city_data_df['Max Temp'] >= max_temp_min + 273.15) &  # Convert to Kelvin
    (city_data_df['Max Temp'] <= max_temp_max + 273.15) &  # Convert to Kelvin
    (city_data_df['Humidity'] <= max_humidity) & #Memeriksa apakah kelembapan kota lebih kecil atau sama dengan kelembapan maksimum yang ditentukan
    (city_data_df['Cloudiness'] == max_cloudiness) & #Memeriksa apakah keterbatasan awan kota sama dengan nilai maksimum yang ditentukan
    (city_data_df['Wind Speed'] <= max_wind_speed) #Memeriksa apakah kecepatan angin kota lebih kecil atau sama dengan kecepatan angin maksimum yang diinginkan.
]

# Drop any rows with null values
filtered_cities = filtered_cities.dropna()

# Display sample data
print(filtered_cities)
```

### Step 3: Create a new DataFrame called `hotel_df`.

```python
# Step 3: Create a new DataFrame called hotel_df
# Use the Pandas copy function to create DataFrame called hotel_df to store the city, country, coordinates, and humidity
hotel_df = filtered_cities[['City', 'Country', 'Lat', 'Lng', 'Humidity']].copy()

# Add an empty column, "Hotel Name," to the DataFrame so you can store the hotel found using the Geoapify API
hotel_df['Hotel Name'] = ""

# Display sample data
print(hotel_df)
```

### Step 4: For each city, use the Geoapify API to find the first hotel located within 10,000 metres of your coordinates.

```python
import requests

# Set parameters to search for a hotel
radius = 10000  # 10,000 meters
params = {
    "apiKey": "d29768fe7b214fad92f045f08f7996f0",  # Replace with your Geoapify API key
    "limit": 1,  # We only want the first hotel
    "categories": "accommodation.hotel"  # Filter for hotels
}

# Print a message to follow up the hotel search
print("Starting hotel search")

# Iterate through the hotel_df DataFrame
for index, row in hotel_df.iterrows():
    # Get latitude and longitude from the DataFrame
    latitude = row["Lat"]
    longitude = row["Lng"]

    # Add the current city's latitude and longitude to the params dictionary
    params["filter"] = f"circle:{longitude},{latitude},{radius}"
    params["bias"] = f"proximity:{longitude},{latitude}"

    # Set base URL
    base_url = "https://api.geoapify.com/v2/places"

    # Make an API request using the params dictionary
    response = requests.get(base_url, params=params)

    # Convert the API response to JSON format
    name_address = response.json()

    # Grab the first hotel from the results and store the name in the hotel_df DataFrame
    try:
        hotel_df.loc[index, "Hotel Name"] = name_address["features"][0]["properties"]["name"]
    except (KeyError, IndexError):
        # If no hotel is found, set the hotel name as "No hotel found".
        hotel_df.loc[index, "Hotel Name"] = "No hotel found"

    # Log the search results
    print(f"{hotel_df.loc[index, 'City']} - nearest hotel: {hotel_df.loc[index, 'Hotel Name']}")

# Display sample data
hotel_df
```

### Step 5: Add the hotel name and the country as additional information in the hover message for each city in the map.

```python
import folium
import pandas as pd
import random

# Function to generate a random color
def random_color():
    return "#{:06x}".format(random.randint(0, 0xFFFFFF))

# Dictionary to store colors for each hotel
hotel_colors = {}

# Create a map centered around the average coordinates
map_center = [hotel_df['Lat'].mean(), hotel_df['Lng'].mean()]
city_map = folium.Map(location=map_center, zoom_start=2)

# Add a point for each city in the DataFrame
for index, row in hotel_df.iterrows():
    # Create a popup message that includes city, country, and hotel name
    popup_message = (
        f"City: {row['City']}<br>"
        f"Country: {row['Country']}<br>"
        f"Hotel: {row['Hotel Name']}"
    )

    # Set a unique color for each hotel, or use grey if no hotel is found
    if row['Hotel Name'] != "No hotel found":
        if row['Hotel Name'] not in hotel_colors:
            hotel_colors[row['Hotel Name']] = random_color()  # Generate a color only once
        marker_color = hotel_colors[row['Hotel Name']]
    else:
        marker_color = 'grey'

    # Size of the marker will be proportional to the humidity
    folium.CircleMarker(
        location=[row['Lat'], row['Lng']],
        radius=row['Humidity'] / 10,  # Adjust the size based on humidity
        popup=popup_message,
        color=marker_color,
        fill=True,
        fill_color=marker_color,
        fill_opacity=0.6
    ).add_to(city_map)

# Create a legend for the hotel names
legend_html = """
    <div style="position: fixed;
                top: 10px; right: 10px; width: 250px; height: auto;
                border:2px solid grey; z-index:9999; font-size:14px;
                background-color:white; padding: 10px;">
        <p style="text-align:center; font-weight: bold;">Hotel Legend</p>
        <ul style="list-style-type: none; padding-left: 0;">
"""

# Add each hotel name to the legend with its corresponding color
for hotel_name, color in hotel_colors.items():
    legend_html += f'<li><span style="color:{color}; font-size:20px;">&bull;</span> {hotel_name}</li>'

legend_html += """
        </ul>
    </div>
"""

# Add the legend to the map
city_map.get_root().html.add_child(folium.Element(legend_html))

# Display the map
city_map
```
