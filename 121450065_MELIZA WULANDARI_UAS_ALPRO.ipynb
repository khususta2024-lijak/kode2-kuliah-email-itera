1. Buatlah fungsi untuk mengubah sebuah string menjadi string lain dengan
aturan sebagai berikut:

a. Hasil result string seluruhnya adalah capital.

b. Tidak ada huruf yang duplikat dalam string resullt.

c. Posisi character string baru tersebut terurut secara ascending, namun index
genap dan ganjil di tukar dalam pair. Contoh: AEJK menjadi EAKJ.

Test Case:
- Input : ‘Institut Teknologi Sumatera’ , output: 'EAIGLKNMROTSU'
- Input: ‘Program Studi Sains Data’ , output: ‘DAIGNMPOSRUT’
- Input: ‘Mata Kuliah Algoritma Pemrograman’ , output: ‘EAHGKIMLONRPUT’

```python
masukkan = input('Masukkan teks yang diinginkan > ')
def swap(nums):
    for i in range(0, len(nums) - 1, 2):
        nums[i], nums[i + 1] = nums[i + 1], nums[i]
    return nums
anagram=''
word = masukkan.upper().replace(" ","")
sorted_word = sorted(set(word))
sorted_word = ''.join(sorted_word)
arr=list(sorted_word)
for x in swap(arr):
    anagram += x
print("Teks yang dimasukkan : "+masukkan)
print("Teks saat sudah menjadi anagram :" + anagram)
```

```python
string = "Program Studi Sains Data"

print('Original Text\t: ', string)

upper = string.upper()
a = list(set(upper))
a.remove(' ')
a.sort()

print(" ")
print('Ascending Huruf tanpa Duplikat\t: ' ,(a))

class liza:
  def solve (self,nums):
    length = len(a)
    for i in range(0, length, 4):
      if (i+1<length):
        a[i], a[i+1] = a[i+1], a[i]
      if (i+2<length):
        a[i+2], a[i+3] = a[i+3], a[i+2]
    return a

ob = liza()
nums = a
ob.solve(nums)
print('Result Text \t\t\t: ', str.join('',nums))
```

```python
kata3 = 'Mata Kuliah Algoritma Pemrograman'
print('Original Text\t: ', kata3)
upper = kata3.upper()
a = list(set(upper))
a.remove(' ')
a.sort()

print(" ")
print('Ascending Huruf\t: ' ,(a))

class huruf:
  def solve (self,nums):
    length = len(a)
    for i in range(0, length, 4):
      if (i+1<length):
        a[i], a[i+1] = a[i+1], a[i]
      if (i+2<length):
        a[i+2], a[i+3] = a[i+3], a[i+2]
    return a

ob = huruf()
nums = a
ob.solve(nums)
print('Result Text \t: ', str.join('',nums))
```
