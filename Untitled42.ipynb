```python
import librosa
import matplotlib.pyplot as plt
import IPython.display as ipd
import numpy as np

# Load audio file
data1, rate = librosa.load("/content/ytmp3free.cc_hindia-membasuh-lyrics-youtubemp3free.org.mp3")

# Function to plot the audio data with time on the x-axis
def show_data(data1, rate):
    plt.figure(figsize=(12, 4))
    plt.title("Waveform of sebelum 1")
    plt.xlabel("Time (seconds)")
    plt.ylabel("Amplitude")
    time = np.arange(len(data1)) / rate  # Convert samples to time in seconds
    plt.plot(time, data1)
    plt.xlim(0, len(data1) / rate)  # Limit x-axis to the duration of the audio
    plt.grid()
    plt.show()

# Display the waveform
show_data(data1, rate)


# Play the audio
ipd.Audio(data1, rate=rate)
```
