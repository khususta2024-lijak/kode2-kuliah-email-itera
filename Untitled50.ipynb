```python
!pip install mysql-connector-python

import mysql.connector
import os

# Koneksi ke database
db = mysql.connector.connect(
    host="localhost",          # Ganti jika pakai remote DB
    user="root",               # Ganti sesuai user MySQL kamu
    password="",    # Ganti dengan password MySQL kamu
    database="database_image"         # Ganti dengan nama database kamu
)

cursor = db.cursor()

# Buka file gambar dalam mode binary
with open('/example.jpg', 'rb') as f:
    binary_data = f.read()

# Query untuk insert data
query = "INSERT INTO image_store (image_name, image_data) VALUES (%s, %s)"
cursor.execute(query, ('example.jpg', binary_data))
db.commit()

print("Image uploaded successfully.")
# Close the connection
cursor.close()
db.close()
```
