```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
import matplotlib.pyplot as plt
```

```python
#import data
df = pd.read_csv('/content/4fc470c4-f6dc-4823-ab08-33a50e570838.csv')
df
```

```python
df = df.drop(['_id', 'periode_data', 'stasiun', 'parameter_pencemar_kritis', 'kategori'], axis=1)
df
```

```python

# Hapus baris yang memiliki tanda '-'
df = df[~df.isin(['-', '---']).any(axis=1)]
df
```

```python
data = df.filter(['max'])
df['tanggal'] = pd.to_datetime(df['tanggal'])
df.set_index('tanggal', inplace=True)
```

```python
#konvert dataframe ke numpy array
dataset = data.values
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(dataset)
```

```python
#split dataset
train_ratio = 0.8
training_data_len = int(np.round(train_ratio * len(dataset)))
```

```python
lookback = 10 #menggunakan 10 hari data sebelumnya untuk prediksi hari setelahnya
x_train, y_train = [], []

for i in range (lookback, len(scaled_data)):
  x_train.append(scaled_data[i-lookback:i, 0])
  y_train.append(scaled_data[i, 0])

#konvert ke numpy array
x_train, y_train = np.array(x_train), np.array(y_train)

x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

```python
model = Sequential()
model.add(LSTM(units=25, return_sequences=True, input_shape=(x_train.shape[1], 1))) #layer 1
model.add(Dropout(0.2)) # layer 2 regulasi untuk overfitting
model.add(LSTM(25, return_sequences=False)) # layer 3 untuk proses data sekuensial
model.add(Dropout(0.2)) # layer 4 dropout untuk regulasi
model.add(Dense(25))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')

history = model.fit(x_train, y_train, batch_size=32, epochs=5, validation_split=0.2)
```

```python
test_data = scaled_data[training_data_len - lookback: , :]

x_test = []
y_test = dataset[training_data_len:, :]
for i in range(lookback, len(test_data)):
  x_test.append(test_data[i-lookback:i, 0])

x_test = np.array(x_test)

x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
```

```python
print(df.dtypes)
```

```python
predictions = model.predict(x_test)
predictions = scaler.inverse_transform(predictions)

# Jika predictions dan y_test adalah array NumPy
predictions = predictions.astype(float)
y_test = y_test.astype(float)


rmse = np.sqrt(np.mean(((predictions - y_test) ** 2)))
mae = np.mean(np.abs(predictions - y_test))
mape = np.mean(np.abs((predictions - y_test) / y_test)) * 100

print("RMSE:", rmse)
print("MAE:", mae)
print("MAPE:", mape)
```

```python
train = df[:training_data_len]
valid = df[training_data_len:]
valid['Predictions'] = predictions

plt.figure(figsize=(16,8))
plt.title('Model')
plt.xlabel('max')
plt.ylabel('tanggal', fontsize=18)
plt.plot(train.index, train['max'])
plt.plot(valid.index, valid[['Predictions', 'max']])
plt.legend(['Train', 'Valid', 'Predictions'], loc='lower right')
plt.show()
```

```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Baca dataset
data = pd.read_csv('/content/4fc470c4-f6dc-4823-ab08-33a50e570838.csv')

# Ubah kolom 'tanggal' menjadi datetime dan urutkan data
data['tanggal'] = pd.to_datetime(data['tanggal'])
data = data.sort_values(by='tanggal')

# Pilih kolom yang relevan untuk prediksi, misalnya 'pm_sepuluh'
data_for_prediction = data[['tanggal', 'pm_sepuluh']]  # Anda bisa mengganti kolom sesuai dengan yang ingin diprediksi

# Normalisasi data
scaler = MinMaxScaler()
data_for_prediction['pm_sepuluh'] = scaler.fit_transform(data_for_prediction[['pm_sepuluh']])

# Siapkan data untuk RNN
def create_sequences(data, seq_length=6):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i + seq_length])
        y.append(data[i + seq_length])
    return np.array(X), np.array(y)

sequence_length = 6  # Panjang sequence (jumlah hari sebelumnya yang digunakan untuk memprediksi)
X, y = create_sequences(data_for_prediction['Scaled_pm_sepuluh'].values, sequence_length)

# Ubah bentuk X agar sesuai dengan input RNN (samples, timesteps, features)
X = X.reshape((X.shape[0], X.shape[1], 1))

# Bangun model RNN dengan lebih banyak layer
model = Sequential([
    SimpleRNN(100, activation='relu', return_sequences=True, input_shape=(sequence_length, 1)),
    SimpleRNN(50, activation='relu'),
    Dense(1)
])
model.compile(optimizer='adam', loss='mse')

# Latih model dengan epoch yang lebih banyak
model.fit(X, y, epochs=200, verbose=0)

# Prediksi nilai berikutnya
last_sequence = data_for_prediction['Scaled_pm_sepuluh'].values[-sequence_length:]  # Data 6 hari terakhir
last_sequence = last_sequence.reshape((1, sequence_length, 1))  # Sesuaikan bentuk input

next_day_prediction_scaled = model.predict(last_sequence)
next_day_prediction = scaler.inverse_transform(next_day_prediction_scaled)  # Kembalikan ke skala asli

# Tampilkan hasil prediksi dalam angka
print("Prediksi PM10 untuk hari berikutnya:", next_day_prediction[0, 0])

# Data prediksi untuk plot
predicted_values = model.predict(X)
predicted_values_rescaled = scaler.inverse_transform(predicted_values)

# Plot data aktual dan prediksi
plt.figure(figsize=(10, 6))
plt.plot(data_for_prediction['tanggal'][sequence_length:], data_for_prediction['pm_sepuluh'][sequence_length:], label='Data Aktual')
plt.plot(data_for_prediction['tanggal'][sequence_length:], predicted_values_rescaled, label='Prediksi RNN')
plt.xlabel('Tanggal')
plt.ylabel('PM10')
plt.title('Prediksi PM10 Menggunakan RNN')
plt.legend()
plt.show()

# Hitung nilai MSE dan MAE
mse = mean_squared_error(data_for_prediction['pm_sepuluh'][sequence_length:], predicted_values_rescaled)
mae = mean_absolute_error(data_for_prediction['pm_sepuluh'][sequence_length:], predicted_values_rescaled)

# Hitung MAPE
actual = data_for_prediction['pm_sepuluh'][sequence_length:].values
predicted = predicted_values_rescaled[:, 0]
mape = np.mean(np.abs((actual - predicted) / actual)) * 100

print(f"Mean Squared Error (MSE): {mse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
```
